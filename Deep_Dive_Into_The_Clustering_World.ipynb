### a.1.2: K-MEANS CLUSTERING: 
Let's start from where we leave out: **Clustering**. Clustering is a very common form of **Unsupervised Learning**, whose goal is to assign similar data points to the same cluster. 

We can formalize it by having a look at what it wants in **Input** and what it returns in **Output**:
- **Input**:
    1. $\quad$ A **Set of Points** $\{ {x}_1, \ldots, {x_n}\}$ in a d-dim vector space ${x} \in \mathbb{R}^d$. 
    2. $\quad$ $k$ **Number of Clusters** to individuate.

- **Output**:
    1. $\quad$ A set of **Centroids** $\{ {\mu}_1, \ldots, {\mu}_k\}$ where ${\mu} \in \mathbb{R}^d$ (Points), and, in particular, one for each Cluster (that needs to be disjoint each other.
    2. $\quad$ A set of **Labels** that create a **mapping from data points to centroids** $\{ y_1, \ldots, y_n\}$.

#### CLUSTERING DIVISION:

There are different approaches and algorithms to perform **Clustering** tasks which can be divided into **three sub-categories**:
- **Partition-Based Clustering**: In this category there are algorithms such as **K-Means**, **K-Medians** and **K-Medoids**. The clustering algorithms of this family create a **Partition of the observations** by **minimizing a certain cost function**.
- **Density-based Clustering**: In this category there are algorithms such as **DBSCAN**. In this category of Clustering, the grouping takes place by analyzing the neighborhood of each point of the space. In particular, the density of points in a neighborhood with a fixed radius is considered.
- **Hierarchical Clustering**: In this category there are algorithms such as **Single Link**, **Average Link** and  **Complete Link**. Hierarchical clustering techniques produce a **Hierarchical Tree Representation**. They are divided into **Agglomerative** and **Divisive** Algorithms. 


#### a.1.2.1: K-MEANS CLUSTERING: ELBOW METHODüí™
A fundamental step for any **Unsupervised Algorithm** is to determine the **Optimal Number of Clusters** into which the data may be clustered. It can be considered, if we do not have any indication, as a sort of **Pre-Processing** step.

(Note that here we are not in a True Unsupervised Algorithm, since we have also Labels. However, we will not use them, up until we will compute Performances. For this reason, we could also forget about them, for now).

The **Elbow Method** is one of the most popular methods useful to determine this **optimal value of k** (number of clusters). In order to choose what is the best numbr of clusters, you have to see at the point of inflection of the curve ("Elbow"). If you use the methodology that includes also the Time, you should opt for the number of clusters that has the best **Distortion - Time** Trade-off. 

The most used metric (Cost Function) is the **Distortion**, which computes the sum of squared distances from each point to its assigned center. $$ \mathcal{J}({\mu},y;{D}) = \sum_{i=1}^N \left| \left| {x}_i - {\mu}_{y_i} \right|\right|^2 $$

```python
#CREATION OF A NEW DATASET:
df1 = df.copy()                                                     #Here we copy the Processor's Dataset.
df1.drop('data', axis=1, inplace=True)                              #We drop from the copy we have just done, all the rows, except for the 'idle_time%'
df1.drop('interrupt_time%', axis=1, inplace=True)                   #column and the 'processor_time%' one.
df1.drop('priviledged_time%', axis=1, inplace=True)
df1.drop('user_time%', axis=1, inplace=True)
df1.drop('Interrupt/sec', axis=1, inplace=True)
df1.drop('state', axis=1, inplace=True)

#ELBOW METHOD:
distortions = []                                                    #Initializing the list "distortions", that will contain all the Distortion Values in it.
K = range(1,10)                                                     #Setting the range of Ks to try pertaining to a range from 1 to 10.
for k in K:                                                         #Looping over the ks.
    kmeanModel = KMeans(n_clusters=k)                               #Applying the Kmeans with k number of clusters.
    kmeanModel.fit(df1)                                             #Fit the data in the K-Means with K centroids.
    distortions.append(kmeanModel.inertia_)                         #Compute and append the Distortion value in the distortions list.
    
#PLOTTING THE ELBOW METHOD RESULT:
plt.plot(K, distortions, 'bx-')                                     #Plotting the Elbow Methos, with K Clusters and the Distortions.
plt.xlabel('k')                                                     #Inserting the x-axis label.
plt.ylabel('Distortion')                                            #Inserting the y-axis label.
plt.title('Distortion: The Elbow Method')                           #Inserting the title in the plot.
plt.show()                                                          #Showing the result.

#ELBOW METHOD WITH VISUALIZER: DISTORTION VS TIME: 
model = KMeans()                                                    #Instantiating the KMeans Clustering.
visualizer = KElbowVisualizer(model, k=(1,10))                      #Creation of the Elbow Visualizer.
visualizer.fit(df1)                                                 #Fitting the data to the visualizer.
visualizer.show()                                                   #Finalizing and render the figure.
```

#### a.1.2.2: K-MEANS CLUSTERING: SILHOUETTE COEFFICIENT


Another method that is commonly used to evaluate the appropriate **Number of Clusters** is the **Silhouette Coefficient**.It is often used as complementary evaluation technique with the elbow Method rather than one being preferred over the other.

The **Silhouette Coefficient** is a measure of **Cluster Cohesion and Separation**. Loosely speaking, is used to evaluate the **Quality of Clusters** in terms of how well samples are clustered with other samples that are similar to each other. It is used to measure **how dense and well-separated the clusters are**. 

It quantifies how well a data point fits into its assigned cluster based on two factors:
- How close the data point is to other points in the cluster.
- How far away the data point is from points in other clusters.

The **Silhouette Coefficient ranges between -1 and 1**. The larger the Silhouette Coefficient, the closer samples are to their cluster w.r.t. they are to other clusters. We have that: 
- If the score is 1, the cluster is dense and well-separated than other clusters. 
- A value near 0 represents overlapping clusters with samples very close to the decision boundary of the neighbouring clusters.
- A negative score [-1, 0) indicate that the samples might have got assigned to the wrong clusters.



The **Silhouette Coefficient** is calculated for each sample of different clusters. In order to calculate the Silhouette score for each observation / data point, the following distances need to be found out for each observations belonging to all the clusters:

‚Ä¢ **Mean Intra-Cluster Distance:** Mean distance between the observation and all other data points in the same cluster. It is denoted by ***a***.

‚Ä¢ **Mean Nearest-Cluster Distance:** Mean distance between the observation and all other data points of the next nearest cluster. It is denoted by ***b***.

The **Silhouette Coefficient S**, for each sample is calculated using the following formula:

$$S = \frac{(b-a)}{max(a,b)} = \frac{Mean Nearest Cluster Distance - Mean Intra Cluster Distance}{max(Mean Intra Cluster Distance, Mean Nearest Cluster Distance)}$$


For instance, if we apply it to my **"Performance Monitoring"** Project, we would obtain that the highest score obtained is at n_clusters = 2 ( about 90%). Here, in the following slide, I will apply it to the basic K-Means version, using Lloyd Algorithm and Random Initialization. 

```python
#DECLARING A LIST OF SILHOUETTE COEFFICIENTS: 
silhouette_coefficients = []                                                           #Declaring an empty list of silhouette coefficients.

#COMPUTING THE SILHOUETTE COEFFICIENTS FOR CLUSTERING IN RANGE [2, 10]:
for k in range(2, 11):                                                                 #For each n_cluster in range [2,10].
   kmeans = KMeans(n_clusters=k, algorithm = 'lloyd', random_state=42, init='random')  #Instantiating the K-Means clustering with k clusters.
   kmeans.fit(df1)                                                                     #Fitting the model.
   score = silhouette_score(df1, kmeans.labels_)                                       #Computing the Silhouette Score.
   silhouette_coefficients.append(score)                                               #Append the result in the list.

#PRINTING OUT THE SILHOUETTE COEFFICIENTS:
print(silhouette_coefficients)                                                         #Printing out the list.

#PLOTTING THE RESULT:
plt.plot(range(2, 11), silhouette_coefficients)                                        #Plotting the silhouette Scores.
plt.title("Silhouette Coefficient K-Means Lloyd Algorithm Random Initialization")      #Setting the title.
plt.xlabel("Number of Clusters")                                                       #Setting the x-label.
plt.ylabel("Silhouette Coefficient")                                                   #Setting the y-label.
plt.show()                                                                             #Showing the plot.
```

#### a.1.2.3: K-MEANS CLUSTERING: RANDOM INITIALIZATION AND LLOYD ALGORITHM:
After having deduced from the Elbow Method, that for our case, the best choice for the **Nummber of Clusters = 2**, we go the Algorithm application. But first...let's recall what is **K-Means**, that can be categorized as a **Partition-Based Clustering**.

The most famous Clustering Algorithm is the **K-Means Lloyd** Algorithm, that uses a **Deterministic** Approach in order to find Samples Labels. Let's now directly go to its 3 main steps:
1. **Centroids Initialization:** We start from some **"Reference" Centroids** $\{ {\mu}_{1}, \ldots, {\mu}_{n}\}$, that are random sampled. (For the Random Sampling, we can use the Trick of Sampling from Available Points). 
2. **Assignment Step:** For every point $i=[1,\ldots,N]$ compute the assignment w.r.t. to the $K$ clusters. 
    - Given a point ${x}_{i}$, **we have to find the label of the "Closest" centroids in $\{ {\mu}_{1}, \ldots, {\mu}_{n}\}$**
        -  $y_i =  \arg\min_k || {x}_{i} - {\mu}_{k} ||_2^2 $  $\quad\quad$ (Here the use of ${L}_{2}$)
3. **Update Step:** Now, we focus on the **Centroids**:
    - For every centroid $k=[1,\ldots,K]$, we **Update** its **Mean depending on the Previous Assignments:**  
        - ${\mu}_{k} \leftarrow \frac{\sum_i \delta\{y_{i}=k\} {x}_{i}}{\sum_i \delta\{y_i=k\}}$  $\quad\quad$ (where $
\delta_{ij} = \begin{cases}
0 &\text{if } i \neq j,   \\
1 &\text{if } i=j.   \end{cases} \quad \text{a.k.a. 1-of-k Coding Scheme} )
$

These clusters that are formed are to minimise the sum of squared distances, and we can view this as finding a **Local Minimum** of the **Distortion Cost Function**. Remember that we are guaranteed that **K-means will always converge** yet to a **Local Optimum**.  The K-means algorithm aims to choose centroids that **minimise Inertia**, or within-cluster sum-of-squares criterion.

#### a.1.2.4: K-MEANS CLUSTERING: RANDOM INITIALIZATION AND ELKAN ALGORITHM:
The **Elkan Algorithm** is an optimized version of the standard K-Means method, based on the fact that most distance calculations in standard **K-Means** are redundant. If a point is far away from a center, it is not necessary to calculate the exact distance between the point and the center in order to know that the point should not be assigned to this center. 
Conversely, if a point is much closer to one center than to any other, calculating exact distances is not necessary to know that the point should be assigned to the first center. 
Indeed, points, usually stay in the same clusters after a few iterations, much of this work is unnecessary, making the Naive implementation very inefficient.

Also, towards the **end of algorithm**, during convergence, there are **very small changes** in the center-assignments of the data points. 

The **Elkan Algorithm** makes use of the **Triangle Inequality**, that is the only property that all distance metrics d possess, that affirms: 

**LEMMA 1: "If center c‚Äô is close to point x, and another center c is far away from c‚Äô, then c‚Äô must be closer to the point x w.r.t. the center c."** <p style="text-align:right;"><img width='20%' src="figs/triangle_inequality.png" />


**COROLLARY:** Some point-center distances of the k-means algorithm do not need to be computed. 

 <p style="text-align:right;"><img width='20%' src="figs/triangle_inequality.png" />


The only **Problem / Drawback** of this technique, is that it requires a lot of memory to store each k lower bounds.
    
Further Readings: Hamerly‚Äôs Algorithm and Drake‚Äôs Algorithm.

### a.1.3: K-MEANS CLUSTERING: K-MEANS++ INITIALIZATION
In addition to the **Optimal Number of Clusters** to choose in a **K-Means Algorithm**, that we could solve through the **Elbow Method** or the **Silhouette Coefficient**, the other Big Deal that we need to deal with is the **Centroid Initialization**, that must be such that the initial cluster centers are placed as close as possible to the optimal cluster centers. 

Indeed, **K-Means Convergence** is only guaranteed to a **Local Optimum**, meaning that with different initializations you end up with different results, that could be very distant from the true **Global Optimum**. Therefore, we have that the **Initialization is  very Important**.

- We have seen the **Random Initialization** that is the way used by the **K-Means Lloyd Algorithm**. We can either select **Initial Centroids completely at random**, or by random sampling from data points we have.
- Another approach could be to use **Multiple Restarts**, i.e., we run the algorithm multiple times from different random starting points, and then pick the best solution. However, this can be slow. Indeed, depending on the different Random Samplings, the K-Means result will converge in different ways (different Local Minima) and speeds. 
- The **Furthest-First Heuristic** idea, is to sample the first Centroid always randomly, but the others will be determined Deterministically  by **Picking Initial Means as far from each other as possible**. However, you do not want to take the **Maximum** across points, due to the **Vulnerability w.r.t. Outliers**. Therefore, with the **K-Means++ Initialization**, we Sample randomly but proportionally according to the distance w.r.t. the centroids. 
$$Pr[{\mu}_k={x}_m] \propto \min_{k<k^{\prime}} \left| \left| {x}_m - {\mu}_k^{\prime}\right|\right|_2^2$$

In K-Means++, higher distances are more likely to be selected.

We can **Implement K-Means++ Initialization** by developing these two steps:
- **Step 1:** Transform the closed set $D$ of distances into a distribution (probability mass function - pdf) as
$$p(d^{\prime})= \frac{d^{\prime}}{\sum_{d \in D} d}  $$
- **Step 2:** We now compute the **Cumulative Sum**(Discrete Case), otherwise do the **Integration**(Continuos Case). Invert this, by Sampling Points from an Uniform Distribution. You have then to find the index by projecting it on the x-axis, finding to what bin it pertains.

The **K-Means Algorithm with K-Means++ Inizialization**, therefore, can be formally said as:
1. Choose ${\mu}_1$ arbitrarily between data points.
2. For $k=[2,\ldots,K]$:
    - **Inverse Transform Sampling** wrt to distances between centroids.
    $$Pr[{\mu}_k={x}_m] \propto \min_{k<k^{\prime}} \left| \left| {x}_m - {\mu}_k^{\prime}\right|\right|_2^2 \quad$$
3. Repeat until convergence Lloyd‚Äôs method:
    - **Assignment step:** $$\forall i\in[1,N] \quad y_i =  \arg\min_k || {x}_i -{\mu}_k ||_2^2 $$
    - **Update step:** $$\forall k\in[1,K] \quad  {\mu}_k \leftarrow \frac{\sum_i \delta\{y_i=k\}{x}_i}{\sum_i \delta\{y_i=k\}}$$

```python
#APPLYING K-MEANS++ CHOOSING THE K DEPENDING ON THE ELBOW METHOD HEURISTIC: 
km_plus = KMeans(n_clusters=2, init='k-means++',algorithm = 'lloyd', random_state=42)             #I select 2 clusters(taking this value from the Elbow Method), and I apply the K-Means ++ Initialization.
klabs_kmeans_plus_lloyd = km_plus.fit_predict(df1)                                                #Saving the Labeling of the K-Means++.

plt.scatter(df['idle_time%'], df['processor_time%'], c=klabs_kmeans_plus_lloyd, s=40, cmap='viridis',)#Plotting the Dataset's Samples, colored thanks to Labeling.
plt.scatter(km_plus.cluster_centers_[:, 0], km_plus.cluster_centers_[:, 1], s=250, marker='*',        #Plotting also the Centroids for every K-Means centroid I have.
            c='red', label='centroids')                                                     
plt.xlabel('processor_time%')                                                               #Inserting the x-axis label.
plt.ylabel('idle_time%')                                                                    #Inserting the y-axis label.
plt.title('KMeans: K-Means++ Initialization and Lloyd Algorithm')                                #Inserting the title in the plot.
plt.legend()                                                                                #Inserting the Legend.
plt.show()                                                                                  #Showing the result obtained.
```

```python
#SETTING UP THE SEED:
np.random.seed(42)                                                      #I set up the seed in order to guarantee reproducibility.   

#SETTING UP THE NUMBER OF CENTROIDS:
N_centers = 2

#STEP 0:
def init(X, K, plusplus=True):                                          #Definition of the "Init" function that will be the one that handles the Initialization, both of the first centroid and both after.
    N, dim = X.shape                                                    #Taking the number of samples and its dimension from the Dataset.
    
    #FIRST CENTROID'S INITIALIZATION:
    if not plusplus:                                                    #When I am in the first initialization of the First Centroid.                                   
        idx = np.random.choice(N, K)                                    #I select k samples randomly from the points I have. 
        return X[idx]                                                   #Return the index of the sample I have randomly choosen as first Centroid.
    
    #OTHERS CENTROID'S INITIALIZATION:
    else:
        idx_first = np.random.choice(N, 1)                              #I select a point.
        centers = X[idx_first]                                          #Initialize the centers. 
        dists = distance(X, centers)                                    #Initialize the distances.
        for k in range(1, K):                                           #Entering the loop from 0 to k-1.
            min_dist = dists.min(axis=1)                                #Compute min across all k-centers, for all points in my Dataset. I have that 'dists' is NxK so min_dist will be Nx1.
            idx = inverse_sampling(min_dist/min_dist.sum(), 1)          #Select one of the N proportional to higher distance.
            new_center = X[idx]                                         #Take its value using the index I am gien above.
            dist_new = distance(X, new_center)                          #Compute the distance between all points to the new center.
            centers = np.vstack([centers, new_center])                  #Update pool of centers.
            dists = np.hstack([dists, dist_new])                        #Update pool of distances.
        return centers                                                  #Return the centers that are computed after the loop.
    

#STEP 1.1:
def inverse_sampling(pmf, n_samples):                                   #Definition of the "Inverse_Sampling" function that will be the one that transforms the closed set D of distances into a distribution
                                                                        #(probability mass function - pdf).
    U = np.random.rand(n_samples)                                       #Take at random n samples.
    sampled_idx = np.argmin((U[:, None] > pmf.cumsum()), axis=1)        #Apply the argmin function in such a way to get the index of the minimum point in terms of distances.
    return sampled_idx                                                  #Return its index.

#STEP 1.2:
def distance(X, centers):                                               #Definition of the "Distance" function that will be the one that that computes the distance between a point and the centroids.
    P = X[:, np.newaxis, :]                                             #Take the Nx1xD data.
    diff_sq = (P - centers)**2                                          #Compute the L2 norm. (Nx1xD - KxD -> NxKxD)
    diff_sq_sum = diff_sq.sum(axis=2)                                   #Compute the square of the L2 Norm. 
    return diff_sq_sum                                                  #Return the sum of the L2 Norm.

#STEP 2.1:
def get_assignments(X, centers):                                        #Definition of the "Get_assignment" function that will be the one that gets the new assignments(labels).
    dist_sq = distance(X, centers)                                      #Get the distance from the points to centroids. 
    y = dist_sq.argmin(axis=1)                                          #Select argmin distance.
    return y, dist_sq.min(axis=1)                                       #Return all the assignments. Y will have an Nx1 Dimensionality.

#STEP 2.2:
def update_centers(X, y, K):                                             #Definition of the "Update_Centers" function that will be the one that handles the Update of centers during my Iterations.
    return np.array([X[y == k].mean(axis=0) for k in range(K)])          #Note that the "==" corresponds to the Delta(i,j) in the formal formula. It corresponds, to the Indicator Function.

#DOUBLE CHECK TO CONTROL WHETHER THE LABELS ARE EQUAL:
centers = init(df_final, N_centers)                                      #Initialize the Centroids with two random numbers.  
y_assignments, _ = get_assignments(df_final, centers)                    #Compute the assignments using the function built above.
labels = pairwise_distances_argmin(df_final, centers)                    #Compute the assignments using the built-in function.
assert np.array_equal(labels, y_assignments), 'Assignment not correct!'  #Check if the result obtained is correct or not.


#WRAPPING UP ALL THE ALGORITHM:
def kmeans_plus_byhand(X, n_clusters):                                  #Function that will merge all the algorithm in one big function.
    np.random.seed(42)                                                  #I set up the seed in order to guarantee reproducibility. 
    iter = 1                                                            #Setting up a counter for counting the number of iterations the algorithm will do.
    
    #STEP 1: RANDOM INIT OF THE CENTROIDS:
    centers = init(X, n_clusters)                                       #Calling the previously contructed function for the Step 1.
    loss = []                                                           #Declare a Loss List for keeping track of the loss I have.
    
    while True:                                                         #Entering the Loop, and therefore entering step 2 and step 3 of the KMeans algorithm.
        
        #STEP 2: ASSIGNMENT STEP:
        y_assignments, dist_sq = get_assignments(X, centers)            #Calling the previously contructed function for the Step 2.
        
        #COMPUTE LOSS:  
        loss.append(dist_sq.mean())                                     #Compute the distance between points and center, average over points then.
        
        #STEP 3: UPDATE STEP:
        new_centers = update_centers(X, y_assignments, n_clusters)      #Calling the previously contructed function for the Step 3.

        #CHECK FOR CONVERGENCE:
        if len(loss) > 2 and np.allclose(loss[-2], loss[-1], atol=1e-5):# If two last losses are close enough I stop.
            break
        
        #UPDATE CENTROIDS FOR THE NEXT ITERATION:
        centers = new_centers                                           #Update the Centroids' new value. 
        iter += 1                                                       #Increase the Iteration number.
        #print(f'> Iteration {iter}: loss {loss[-1]}')                  #Print ou the Loss I have at the i-th iteration.

    return centers, y_assignments                                       #Return from KMeans both the centroids and the labels.


#CALLING THE KMEANS++ BY-HAND IMPLEMENTATION:
centers, labs_kmeans_plus_byhand = kmeans_plus_byhand(df_final, N_centers)#Obtainig the centroids and labels of the KMeans.



#PLOTTING THE KMEANS BY-HAND IMPLEMENTATION RESULT:                
plt.scatter(*df_final.T, c=labs_kmeans_plus_byhand, s=40, cmap='viridis')#Plotting the Dataset's Samples, colored thanks to Labeling.
plt.scatter(*centers.T, s=250, marker='*', c='red', label='centroids')  #Plotting also the Centroids for every K-Means centroid I have. 
plt.xlabel('processor_time%')                                           #Inserting the x-axis label.
plt.ylabel('idle_time%')                                                #Inserting the y-axis label.
plt.title('K-Means++ By Hand Resolution')                               #Inserting the title in the plot.
plt.legend()                                                            #Inserting the Legend.
plt.show()                                                              #Showing the result obtained.
```

### a.1.4: K-MEANS CLUSTERING: NAIVE SHARDING CENTROID INITIALIZATION
The **Sharding Centroid Initialization Algorithm** is another **Initialization Method**. Sharding Initialization is expected to execute much **Quicker** w.r.t. the random centroid initialization, given the fact that, under a Timing perspective, randomly initializing centroids for increasingly complex datasets grows superlinearly. 
Sharding initialization executes in linear time, and is not dependent on data complexity.

The **Sharding Procedure** is the following:

- **Step 1**: Sum up each sample (row) of the Dataset, and save this result on a new column "Sum_Rows". 
- **Step 2**: Sort the Dataset, in ascending order, depending on the newly created column "Sum_Rows". 
- **Step 3**: Split the dataset horizontally into k equal-sized pieces, or **Shards** (k = n_clusters).

<p style="text-align:center;"><img width='30%' src="figs/naive_sharding_1.png" />

- **Step 4**: For each shard, sum the samples vertically, namely for each Attribute (a.k.a. Feature) (excluding the "Sum_Rows" column). After this, compute its Mean, and save the values into a new row that will represent one of the  **Naive Sharding Centroids**.

After this, we can continue with the basic **Lloyd Algorithm**, by passing to it the Centroids we have just calculated.

<p style="text-align:center;"><img width='30%' src="figs/naive_sharding_2.png" />



```python
def Naive_Sharding(dataset, k):
    #COMPUTING THE DATASET's HEIGHT AND WIDTH. 
    n = np.shape(dataset)[1]                                                            #Take teh Width.
    m = np.shape(dataset)[0]                                                            #Take the Height.
    
    #INITIALIZE THE CENTROIDS IN A MATRIX FORM.
    centroids = np.asarray(np.zeros((k,n)))                                             #Initialize all the centroids.

    #STEP 1: SUMMING UP ALL ELEMENTS FOR EACH ROW AND SAVING THE RESULT ON A NEW COLUMN.
    composite = np.mat(np.sum(dataset, axis=1))                                         #Sum up all the elements for each row.
    dataset = np.append(composite.T, dataset, axis=1)                                   #Append the new column.
    
    #STEP 2: SORTING THE DATASET, IN ASCENDING ORDER, DEPENDING ON THE NEWLY CREATED COLUMN.
    dataset.sort(axis=0)                                                                #Sort the Dataset.

    #STEP 3: SPLIT THE DATASET HORIZONTALLY INTO K EQUALLY-SIZED PIECES (SHARDS). 
    step = m // k                                                                       #Compute the Mean.    
    vfunc = np.vectorize(_get_mean)                                                     #Vectorize mean ufunc for numpy array.

    #STEP 4: FOR EACH SHARD, SUM THE SAMPLES VERTICALLY, NAMELY FOR EACH ATTRIBUTE (A.K.A. FEATURE). COMPUTE ITS MEAN AND SAVE THE VALUES INTO A NEW ROW THAT WILL REPRESENT ONE OF
    #        THE NAIVE SHARDING CENTROIDS.
    for j in range(k):                                                                  #For every centroid.
        if j == k-1:                                                                    #If it is not the last Centroid.
            centroids[j:] = vfunc(np.sum(dataset[j*step:,1:], axis=0), step)            #Save the centroid value by computing the whole procedure.
        else:                                                                           #If it is the last centroid.
            centroids[j:] = vfunc(np.sum(dataset[j*step:(j+1)*step,1:], axis=0), step)  #Compute the procedure.

    return centroids                                                                    #Returning the Centroids.

def _get_mean(sums, step):
    return sums/step                                                                    #Returning sums / step. 

#RESTRICTING THE DATASET:
df_naive = df.copy()                                                      #Copying the initial Dataset.                                      
df_naive.drop('data', axis=1, inplace=True)                               #Dropping from the Dataset the 'data' column.
df_naive.drop('interrupt_time%', axis=1, inplace=True)                    #Dropping from the Dataset the 'interrupt_time%' column.
df_naive.drop('priviledged_time%', axis=1, inplace=True)                  #Dropping from the Dataset the 'priviledged_time%' column.
df_naive.drop('user_time%', axis=1, inplace=True)                         #Dropping from the Dataset the 'user_time%' column.
df_naive.drop('Interrupt/sec', axis=1, inplace=True)                      #Dropping from the Dataset the 'Interrupt/sec' column.
df_naive.drop('state', axis=1, inplace=True)                              #Dropping from the Dataset the 'state' column.

#DECLARING THE SCALER: 
minmax_scaler = sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1))  #Declaring the MinMax Scaler.
X_scaled = minmax_scaler.fit_transform(df_naive)                          #Scale the Dataset and computing the Labels.
X_trans_inv = minmax_scaler.inverse_transform(X_scaled)                   #Computing the Inverse Transform in order to get back results not-scaled.


#USING THE NAIVE SHARDING INITIALIZATION: 
centroids_scaled = Naive_Sharding(X_scaled, 2)                            #Computing Centroids using the Scaled Version.
centroids = Naive_Sharding(X_trans_inv, 2)                                #Computing Centroids using the non-Scaled Version.
print(centroids_scaled, centroids)
```

### a.1.5: K-MEDIANS CLUSTERING

Another very important **Clustering Algorithm** is the so called **K-Medians Algorithm**. Compared with the K-Means, they both involve finding K Cluster centers for which the sum of the distance between a center and all points in that cluster is minimized. 

Like the **K-Means Clustering**, also **K-Medians** belong to the **Partition-Based Clustering** category.

##### WHEN TO USE IT:
One of the **K-Means drawbacks** is its **Sensitivity to Outliers**, as it uses and minimizes **L2 Norm** (a.k.a. **Euclidean Distance** as the dissimilarity metric. Unfortunately, real-world data sets often come with many outliers that you might not be able to remove completely during the data cleanup phase. 

The **K-Medians Algorithm** uses the **Median** instead of the mean, and uses and minimizes the way **More Robust** dissimilarity metric, much less sensitive to outliers, the ***L1 Norm*** distance (a.k.a. **Manhattan Distance** or **Taxicab Norm**). 

The **Median**, is a statistic incredibly resistant to outliers because in order to deter the median away from the
bulk of the information, it requires at least 50% of the data to be "contaminated".



The Criterion function used for **K-Medians** is the following:
$$ S = \sum_{k=1}^K \sum_{x_i \in C_k}^K  \left|\left| {x}_i - {Median}_{C_k} \right| \right|_1$$ 

So it is the sum over the number of Clusters, in which for each cluster $C_k$, we are going to sum the absolute distance between the point $x_i$ and the Cluster's Median.  

The **K-Medians Algorithm** can therefore be summurized as:
- Select at the start K points that will be the initial Centroids (Medians).
- Repeat until convergence: 
    - **Assignment step:** For every point $i=[1,\ldots,N]$ compute the assignment w.r.t. to the $K$ clusters. Assign every point to its nearest Cluster (${\mu}_k$). 
    $$\forall i\in[1,N] \quad y_i =  \arg\min_k || {x}_i -{\mu}_k ||_1   \quad\quad({L}_{1})$$
    - **Update step:** Recompute the Median using the Median of each individual Feature.
    $$\forall k\in[1,K] \quad  {\mu}_k \leftarrow Median$$
    where, we have that:
        - Before any step, **Sort the Array of Points for the k-th cluster in Ascending Order**. Then: 
        - If the number of points $n$ is assigned to the k-th Cluster is ***Odd***, we have that the Median corresponds to this term: $$Median = {\frac{n + 1}{2}}$$
        - f the number of points $n$ is assigned to the k-th Cluster is ***Even***, we have that the Median corresponds to this term: $$Median = \frac{(\frac{n}{2}) + (\frac{n}{2} + 1)}{2}$$

##### Ex1:

Given for the k-th Cluster the list of points $l$ = [12, 16, 12, 6, 18, 2, 4], its Median in computed, given the **Odd** number of points, as: 
- Sort in Ascending Order: $l$ = [2, 4, 6, **12**, 12, 16, 18]
- The Median, will be equal to the Middle-Point, namely: $$Median = {\frac{n + 1}{2}} = \frac{7+1}{2} = 4_{th} \rightarrow 12$$ 

##### Ex2:

Given for the k-th Cluster the list of points $l$ = [42, 7, 17, 14, 7, 24, 15, 29], its Median in computed, given the **Even** number of points, as: 
- Sort in Ascending Order: $l$ = [7, 7, 14, **15**, **17**, 24, 29, 42]
- The Median, will be equal to the Middle-Point, namely: $$Median = \frac{(\frac{n}{2}) + (\frac{n}{2} + 1)}{2} = \frac{\frac{8}{2} + (\frac{8}{2}+1)}{2} = \frac{4_{th} + 5_{th}}{2} = \frac{15 + 17}{2} \rightarrow 16$$ 



This is the implementation for the **K-Medians Algorithm with Random Initialization**. 

We are then going to apply it to our **Idle_Time% VS Pcosessor_Time%** Clustering.

```python
#STEP 0:
def init_kmedians(X, k):                                                    #Declaring a funciton for my 0 Step.
    N, dim = X.shape                                                        #Recover the Dimensions of my Dataframe.
    idx = np.random.choice(N, k)                                            #I select k samples from the points I have randomly. Note that actually the indexes are random sampled. 
    return X[idx]                                                           #Since I random sample the indexes, I return the k-random sampled points.

#STEP 1:
def get_assignments_kmedians(X, centers):                                   #Declaring a funciton for my 1 Step.
    P = X[:, np.newaxis, :]                                                 #I have to find the label of the "closest" centroids (Nx1xD).
    diff_abs = np.abs(P - centers)                                          #Computing the L1 Loss. (Nx1xD - KxD -> NxKxD).
    y = diff_abs.sum(axis=2).argmin(axis=1)                                 #L1 Loss applied.
    return y                                                                #Return all the assignments. It will have an Nx1 Dimensionality.

#STEP 2:
def update_centers_kmedians(X, y, K):                                       #Declaring a funciton for my 2 Step.
    # Select the points to the k center and remove N dimension with mean over axis 0.
    return np.array([np.median(X[y == k] ,axis=0) for k in range(K)])       #Note that the "==" corresponds to the Delta(i,j) in the formal formula. It corresponds, to the Indicator Function.

#WRAPPING UP ALL THE ALGORITHM:
def kmedians_byhand(X, n_clusters):                                         #Declaring the K-Medians function taking in input a Dataset and a number of clusters.                                 
    np.random.seed(42)                                                      #I set up the seed in order to guarantee reproducibility. 
    iter = 1                                                                #Setting up a counter for counting the number of iterations the algorithm will do.
    
    #STEP 1: RANDOM INIT OF THE CENTROIDS:
    centers = init_kmedians(X, n_clusters)                                  #Calling the previously constructed function for the Step 1.
    
    while True:                                                             #Entering the Loop, and therefore entering step 2 and step 3 of the KMeans algorithm.
        
        #STEP 2: ASSIGNMENT STEP:
        y_assignments = get_assignments_kmedians(X, centers)                #Calling the previously contructed function for the Step 2.
        
        #STEP 3: UPDATE STEP:
        new_centers = update_centers_kmedians(X, y_assignments, n_clusters) #Calling the previously contructed function for the Step 3.
        
        #CHECK FOR CONVERGENCE:
        if np.allclose(centers, new_centers):                               #This is a basic implementation of this step. A more advanced one would be to check for if the loss is plateauing.
            #print(f'KMeans terminates after the Iteration number: {iter}') #Print out the number of iteration at which the algorithm stops.
            break                                                           #The "np.allclose" function returns True if two arrays are element-wise equal within a tolerance. Break the while if it is so.
        
        #UPDATE CENTROIDS FOR THE NEXT ITERATION:
        centers = new_centers                                               #Update the Centroids' new value. 
        iter += 1                                                           #Increase the Iteration number.
    
    return centers, y_assignments                                           #Return from KM-Medians both the centroids and the labels.
```

### a.1.6: K-MEDOIDS CLUSTERING

Another **Clustering Algorithm** is the so called **K-Medoids**, very similar both to the **K-Means** and to the **K-Medians** Clustering techniques. 

##### K-MEDOIDS VS K-MEANS: 
We already know the very bad **K-Means Clustering Vulnerability**, the **Sensitivity to Outliers**, due to the fact that a Mean is easily influenced by extreme values. **K-medoids Clustering** is a variant of K-means that is more robust to noises and outliers. 

Instead of using the **Mean** point as the center of a cluster (which **may or may not be one of the data points**), K-medoids uses an actual point in the cluster to represent it. A **Medoid** is defined to be the most centrally located object of the cluster, with minimum sum of distances to other points. 

<p style="text-align:center;"><img width='30%' src="figs/mean_medoid.png" />

This example above shows the difference between mean and medoid in a 2-D example. The group of points in the left form a cluster, while the rightmost point is an outlier. Mean is greatly influenced by the outlier and thus cannot represent the correct cluster center, while medoid is robust to the outlier and correctly represents the cluster center.




Besides, **K-Medoids Algorithm** can be used with arbitrarily chosen dissimilarity measure (any distance metric) unlike **K-Means** which usually needs **Euclidean Distance** metric to arrive at efficient solutions.

#### a.1.6.1: K-MEDOIDS VS K-MEDIANS:
**K-Medoids Clustering** is also very similar to the **K-Medians Clustering**, that we know to be, loosely speaking, the same as K-means, except that it chooses the **Medians** (instead of means) of the clusters as centres. However we have that in **K-Medians** we could have:
- **Odd # of Points:** In the case in which we have an **Odd** number of points, it will opt for the Median Data Point, thus always returning an actual **Medoid**. 
(Ex. [1, 2, 3, 6, 7] $\rightarrow$ 3).
- **Even # of Points:** However, in this case, when we have an **Even** number of points, it will opt for the mean between the two internal data points. Here we have two cases that could come out depending on the **two internal data points**:
    - **Same Internal Data Points:** In the case in which the two Internal Data Points are equal, the point to be choosen, will still be an actual **Medoid**, thus obtaining a point that is actually present in the Dataset. 
        (Ex. [1, 2, 2, 6] $\rightarrow$ 2+2 /2 = 2).
    - **Diverse Internal Data Points:** In the case in which the two Internal Data Points are diverse, the point to be choosen, will **Not** be an actual **Medoid**, thus obtaining a point that is **not** actually present in the Dataset. 
        (Ex. [1, 2, 3, 6] $\rightarrow$ 2+3 /2 = 2.5).

### a.1.6.2: K-MEDOIDS CLUSTERING: PARTITIONING AROUND MEDOIDS (PAM)

In general, the **K-Medoids Problem** is **NP-hard** to solve exactly. As such, many heuristic solutions exist. The most common Algorithm used to face the K-Medoids Problem is the **Partitioning Around Medoids (PAM)** algorithm that uses a greedy search which may not find the optimum solution, but it is faster than exhaustive search. (It can be thought of as the Lloyd Algorithm to the K-Means Clustering Problem). 

It works as follows:
- **Build Step (Initialization)**: Select k of the n data points as the Medoids to minimize the cost.
- **Associations Step**:Associate each data point to the closest medoid. 
$$y_i =  \arg\min_k || {x}_{i} - {\mu}_{k} ||_1 \quad\quad (Here \space the \space use \space of {L}_{1})$$
- **Swap Step**: While the cost of the configuration decreases:
    - For each medoid m, and for each non-medoid data point o:
        - Consider the swap of m and o, and compute the **Cost** change. (here, we consider the **Manhattan Distance**)
        $$Cost = \sum_{m_i} \sum_{o_i \in m_i}  \left| {o}_i - {m}_{i} \right|$$
        - If the cost change is the current best, remember this m and o combination.
    - Perform the best swap of $m_{best}$ and $o_{best}$, if it decreases the cost function. Otherwise, the algorithm terminates.

The type of **Initialization** used in the **PAM Algorithm** is known as **Build**". It is a **greedy initialization** of the medoids. 

During the **Build** phase the first medoid is selected to be the one that has the minimum cost, with cost being the sum over all distances to all other points. Therefore, the first point is the most central point of the data set.

All further points are then selected iteratively. For all non-medoids we calculate the cost for selecting this point as the next medoid (again the sum of distances from the candidate medoid to all other non-medoids), and then select the one with the smallest cost as the next medoid.

Often it is **more efficient but slower** than other initializations **on big datasets** and it is also **very non-robust**, if there are outliers in the dataset, use another initialization.

#### a.1.6.3: K-MEDOIDS CLUSTERING: VORONOI ITERATION

Another option for solving the **K-Medoids Problem** is the so called **Voronoi Iteration** method a.k.a. the **Alternating Heuristic** in literature, as it alternates between two optimization steps:
- Select initial medoids randomly.
- Iterate while the cost decreases:
    - In each cluster, find the **point that minimizes the sum of distances** within the cluster, and make it to be the **Medoid**.
    - Reassign each point to the cluster defined by the closest medoid determined in the previous step.

Note that, during the **Initialization Step**, we have to select the initial **Medoids** from the dataset. The **Random Initialization**, here, is not the only choice we have. Indeed, we could opt for:
- **Random Initialization**: It selects randomly the **Medoids** choosing among the Dataset's samples. 
- **Heuristic Initialization**: It selects the **Medoids** picking the points with the smallest sum distance to every other point.
- **K-Medoids++ Initialization**: It selects the **Medoids** by following an approach based on k-means++. In general, it gives initial medoids which are more separated than those generated by the other methods.

(Note that they are all configurable using the **init** parameter).

I will apply this last Algorithm to my **Idle VS Working** example, first using the **K-Medoids Alternate Algorithm with Random Initialization (Manhattan Distance)**, then using the **K-Medoids Alternate Algorithm with Heuristic Initialization (Manhattan Distance)** and, in the end, using the **K-Medoids Alternate Algorithm with K-Medoids++ Initialization (Manhattan Distance)**.

#### a.1.6.4: K-MEDOIDS CLUSTERING: CLARA AND CLARANS (HINT)

Other two approximate algorithms such as **CLARA** and **CLARANS** trade quality for runtime. 
- **CLARA (Clustering Large Applications)**: It is able to deal more efficiently with larger data sets than PAM method. CLARA applies **Sampling Approach** to handle large data sets. Rather than finding medoids for the entire data set D, CLARA first draws a small sample from the data set and then applies the PAM algorithm to generate an optimal set of medoids for the sample. The **quality of resulting medoids** is measured by the average dissimilarity between every item in the entire data space D and the medoid of its cluster. The cost function is defined as follows: $$Cost(m_d, D) = \sum_{i-1}^n   \left| {x}_i - {m}_{i} \right| $$
where we have that $m_d$ is a set of selected Medoids, and $m_i$ is a Medoid in $m_d$ which is closest to $x_i$.
The sampling and clustering processes are repeated a pre-defined number of times.
The clustering that yields the set of medoids with the minimal cost is selected. 


- **CLARANS (Clustering Large Applications based on RANdomized Search, )**: It works on the entire data set, but only explores a subset of the possible swaps of medoids and non-medoids using sampling. It interprets the search space as a high-dimensional hypergraph, where each edge corresponds to swapping a medoid and non-medoid.  CLARANS draws sample of neighbours dynamically. This clustering technique mimics the graph search problem wherein every node is a potential solution, here, a set of k medoids. If the local optimum is found, search for a new local optimum is done with new randomly selected node. It is more efficient and scalable than both PAM and CLARA.

#### a.1.6.6: EXAMPLE: DIGIT CLASSIFICATIONS

Let's now delve into a very Practical Implementation in Python, where I will use the **K-Medoids Algorithm** on a dataset containing 8*8 dimensional images of **Handwritten Digits**. 

##### TASK:
The task consists in dividing the data points into 10 clusters (for classes 0-9) using K-Medoids. 

##### IMPLEMENTATION:
In the implementation below, we will also make use of the **PCA**, in order to bring data in a subdimensional space, and also **Scaling**, in such a way to standardize data and make the process faster. We will use three different metrics: 
- Euclidean Distance
- Manhattan Distance
- Cosine Distance

In such a way to see in practice what is the difference in Clustering terms in using these three different metrics.

```python
#LOADING THE DIGIT DATASET:  
dataset = load_digits()                                                    #Loading the Digit Dataset.

#STANDARDIZING THE DATASET: 
digit_data = scale(dataset.data)                                           #Rescaling the Dataset in usch a way that it has mean of 0 and standard deviation of 1.

#COMPUTING THE NUMBER OF DIGITS:
num_digits = len(np.unique(dataset.target))                                #Compute the number of digits we have (10 in our case).

#APPLYING PCA:
red_data = PCA(n_components=2).fit_transform(digit_data)                   #Instead of using all 64 attributes of the dataset, we use Principal Component Analysis (PCA)  to reduce the dimensions of features set such that most of the useful information is covered.

#PLOTTING THE DECISION BOUNDARIES FOR EACH CLUSTER:
h = 0.02                                                                   #Step size of the mesh. 
xmin, xmax = red_data[:, 0].min() - 1, red_data[:, 0].max() + 1            #Definying minimum and maximum x-coordinates.
ymin, ymax = red_data[:, 1].min() - 1, red_data[:, 1].max() + 1            #Definying minimum and maximum y-coordinates.
xx, yy = np.meshgrid(np.arange(xmin, xmax, h), np.arange(ymin, ymax, h))   #Creation of the Decision Boundaries Points.

#DEFINYING AN ARRAY OF K-MEDOIDS VARIANTS:
models = [
     (KMedoids(metric="manhattan", n_clusters=num_digits, method = 'alternate',
      init="heuristic", max_iter=100),"Manhattan metric"),
     (KMedoids(metric="euclidean", n_clusters=num_digits, method = 'alternate',
      init="heuristic", max_iter=100),"Euclidean metric"),
     (KMedoids(metric="cosine", n_clusters=num_digits, method = 'alternate',
      init="heuristic", max_iter=100), "Cosine metric")] 

#INITIALIZING THE NUMBER O F ROWS AND COLUMNS OF THE PLOT FOR PLOTTING SUBPLOTS:
num_rows = int(np.ceil(len(models) / 2.0))                                 #Defining the Number of Rows.
num_cols = 2                                                               #Defining the Number of Columns.

plt.clf()                                                                  #Clearing the current figure first.                                                  
plt.figure(figsize=(15,10))                                                #Initializing the dimensions of the plot.

#ITERATING OVER THE MODELS:
for i, (model, description) in enumerate(models):                          #For every index and model in models.
    
    #FITTING POINTS IN THE MESH TO THE MODEL.
    model.fit(red_data)                                                    #Fit each point in the mesh to the model.
    
    #LABEL PREDICTION.
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])                       #Predict the labels for points in the mesh.
    
    #BUILDING THE COLOR PLOT.
    Z = Z.reshape(xx.shape)                                                #Put the result  into a color plot.
    
    #CREATION OF A SUBPLOT FOR THE i-th MODEL.
    plt.subplot(num_cols, num_rows, i + 1)                                 #Subplot for the ith model variant.
    
    #SHOWING THE SUBPLOT:                               
    plt.imshow(Z, interpolation="nearest",                                 #Data and Interpolation.
         extent=(xx.min(), xx.max(), yy.min(), yy.max()),                  #Bounding box.
         cmap=plt.cm.Paired,                                               #Colormap.
         aspect="auto",                                                    #Aspect ratio of the axes.
         origin="lower")                                                   #Set origin as lower left corner of the axes.
   
    plt.plot(                                                              #Plot Data.
        red_data[:, 0], red_data[:, 1], "k.", markersize=2, alpha=0.3)
    
    #PLOTTING CENTROIDS: 
    centroids = model.cluster_centers_
    plt.scatter( centroids[:, 0], centroids[:, 1], marker="*", s=169,      #Plotting the centroids.
         linewidths=3, color="white",  zorder=10)
    plt.title(description)  
    plt.xlim(xmin, xmax)                                                   #X-coordinates Limit.
    plt.ylim(ymin, ymax)                                                   #Y-coordinates Limit.                           
    plt.xticks(())                                                         
    plt.yticks(())
    plt.suptitle("K-Medoids algorithm implemented with different metrics")
plt.show() 
```

### a.1.7: MEAN-SHIFT CLUSTERING

Another very important **Clustering Algorithm** is the so called **Mean-Shift Clustering**, that is an unsupervised learning algorithm. It is widely used in real-world data analysis (e.g., **Image Segmentation**)because it‚Äôs **Non-Parametric** and doesn‚Äôt require any predefined shape of the clusters in the feature space.

Loosely speaking, ‚ÄúMean Shift‚Äù means ‚Äúshifting to the mean‚Äù in an iterative way. In the algorithm, every data point is shifting to the ‚Äúregional mean‚Äù step by step and the location of the final destination of each point represents the cluster it belongs to.

The **Mean-Shift** is used for locating the maxima of a density function, and this is the reason why it is a.k.a. **Mode-Seeking Algorithm**.
<img src="https://cdn-images-1.medium.com/max/800/1*bkFlVrrm4HACGfUzeBnErw.gif" width="500" align="center">

**Mean-Shift Algorithm** aims to discover blobs in a smooth density of samples. It is a centroid-based algorithm that works by updating candidates for centroids to be the mean of the points within a given **region** (a.k.a **Bandwidth**). These candidates are then filtered in a post-processing stage to eliminate near-duplicates to form the final set of centroids. 

Therefore contrary to K-Means, **WE DO NOT NEED TO CHOOSE THE NUMBER OF CLUSTERS!**

##### ALGORITHM: 
On a high level, **Mean Shift Algorithm** works as follows:
- **Step 1 (Creation Step)**: Creation of a **Sliding Window/cluster** for each data-point.
- **Step 2 (Shifting Step)**: Each of the **sliding windows is shifted towards higher density regions** by shifting their centroid (center of the sliding window) to the data-points' mean within the sliding window. This step will be **repeated until no shift yields a higher density** (number of points in the sliding window).
- **Step 3 (Post-Processing)**: Here it occurs a **selection of sliding windows** by **deleting overlapping ones**. When multiple sliding windows overlap, the greedy choice is to mantain the window containing the most points, and the others are deleted.
- **Step 4 (Assignment Step)**: Assigning the data points to the sliding window in which they reside.

#### a.1.7.1: MEAN-SHIFT CLUSTERING: INTUITION

What we're trying to achieve during the algorithm is therefore to keep shifting the window to a **Region of Higher Density**. This is why, we keep shifting the window towards the centroid of all the points in the window. This feature of Mean Shift algorithm categorizes it as **Hill Climb Algorithm**.

The first step when applying mean shift is representing your data in a mathematical manner. For mean shift, this means **representing your data as points**, such as the set below.
<p style="text-align:right;"><img width='40%' src="figs/mean12.png" />

**Mean Shift** is built upon the concept of **Kernel Density Estimation (KDE)**. Imagine that the above data was sampled from a probability distribution. **KDE** is a method to **estimate the underlying distribution** (a.k.a. the probability density function) for a set of data.

It works by placing a **Kernel** on each point in the data set. A kernel is a fancy mathematical word for a **weighting function**. There are many different types of kernels, but the most popular one is the Gaussian kernel. Adding all of the individual kernels up generates a probability surface (e.g., density function). Depending on the kernel bandwidth parameter used, the resultant density function will vary.

Below is the KDE surface plot for our points above using a Gaussian kernel with a kernel bandwidth of 2. 
<p style="text-align:right;"><img width='30%' src="figs/mean11.png" />


**Mean-Shift** exploits this KDE idea by imagining what the **points** would do if they all **climbed up hill to the nearest peak on the KDE surface**. It does so by iteratively shifting each point uphill until it reaches a peak.

Depending on the **Kernel Bandwidth** used, the KDE surface (and end clustering) will be different. Let's look at the two extremes:

- A **Super Large Bandwidth (extremely fat Kernel)** setting can disable us from seeing the local structure of the dataset.  This will result in a wide smooth KDE surface with **one peak** that all of the points will climb up to, resulting in one cluster. 

- A **Super Small Bandwidth (extremely tall skinny Kernel)**, there will be a lot more non-sense clusters because we restricted the mean value calculation to a too smaller local area. The resultant **KDE Surface** will have a peak for each point. This will result in **each point being placed into its own cluster**.  As an example of a too small Bandwidth is the following:
<img src="https://spin.atomicobject.com/wp-content/uploads/ms_2d_bw_.8.gif" width="500" align="center">

Kernels in between these two extremes will result in nicer clusterings. Here the example of the perfect choice:
<img src="https://spin.atomicobject.com/wp-content/uploads/ms_2d_bw_2.gif" width="500" align="center">




#### a.1.7.2: MEAN-SHIFT CLUSTERING: A CLOSER LOOK
By having a look at a sample dataset with two features, like this:  
<p style="text-align:right;"><img width='30%' src="figs/mean1.png" />
    
we can define the **Mean-Point** fo the entire Dataset (red one) to be the **Arithmetic Mean** for Feature1 and for Feature2, beacuse it is calculated based on the equal weights to all points: 
    $$M_a = \frac{1}{n}\sum_{i=1}^n x_i$$

where we have $M$ to be the Mean, $n$ to be the sample size, and $x_i$ is one feature (feature_1 or feature_2)of the data points.


However, sometimes it‚Äôs not reasonable to give equal weights to everything. A better approach is to compute the **Weighted Mean**. $$M_w = \frac{\sum_{i=1}^n w_i x_i}{\sum_{i=1}^n w_i}$$

where we have that $w_i$ is the weight for $x_i$ and it is proposrtional to the **Euclidean Distance** between points. The most widely used **Weight Function** in mean shift algorithm is the **Flat Function (Kernel)**, a sort of **Indicator Funciton**:
$$ w(d) = \begin{cases}
1 &\text{if } d \leq R,   \\
0 &\text{if } d > R.   \end{cases}
$$

where $d$ is the distance between any data point to the currently investigated one, and $R$ is the radius of the circle centered at the investigated point.

For instance, if we want to calculate the weighted mean for the area around point $O$, an indicator function can be understood as a **Hard Boundary Circle centered at $O$ with Radius $R$**. And we have that:
- Everything inside the circle will be counted.
- Any data point outside the circle will be ignored.
<p style="text-align:right;"><img width='50%' src="figs/mean2.png" />

#### a.1.7.3: MEAN-SHIFT CLUSTERING: GAUSSIAN IDEA
A very straight comparison is the following: it is like we are standing on one local point (center point O) and cannot see the entire picture but are restricted to a local area to calculate the mean (we do not have a global view of outside. Imagin as if there is some fog and you can only see at a small distance). 

Here, we see the red point is the **Weighted Mean** of the investigated area and we found that it tends to **locate** at a region with **high density of points**.

We can also replace the flat weight function with a **Gaussian** Function:
$$w(d) = e^{-\frac{d}{2\sigma^{2}}}$$

where $d$ is still the distance between the center point to any other point, and $\sigma$ here is a parameter to adjust **how fast the weight decreases** with the increasing of the distance.

The idea is similar to that of the flat weight function (Indicator Function): the closer a point is to the circle center, the higher the weight is in the mean calculation. 

For simplicity of implementation and clear physical meaning, **Mean-Shift Algorithm** typically uses the **Flat Weight Function** to calculate the mean.

The local mean point we calculated above kind of represents the position with the largest density of points in that specific local area. 

Notice that the **Radius of the Black Circle, R**, in the plot above, is **very Important** to define the local region. Actually, the **Radius** is the **only Parameter** in the **Mean-Shift Algorithm**, a.k.a **Bandwidth**.

#### a.1.7.4: MEAN-SHIFT CLUSTERING:  WHY SHIFTING?
The **Shifting Step** (a.k.a. **Sliding Step**) does not change the raw data. The **Shifting** just describes the process of how we label each data point. Similar points gradually ‚Äúshift‚Äù to the centroid of the cluster they belong to, in order to be labeled as such.

###### PRACTICAL EXAMPLE:
Let‚Äôs look at a 2-D toy dataset, which is designed to have three clusters.
<p style="text-align:right;"><img width='40%' src="figs/mean3.png" />
    

When we are going to apply the **Mean Shift Algorithm**, each point try to find its group by moving towards the weighted mean of its local area in each step. The destination of each point will be the centroid of the data cluster that the point belongs to. Then, all the data points with the same destination point can be labeled with the same cluster.
<p style="text-align:right;"><img width='40%' src="figs/mean4.png" />
    
The plot above shows the result of labeling, where the **black big circles** indicate the **final destinations** of each group of data points and the **arrows** approximately shows the **trajectory** of some of the moving paths.

Therefore, as you can see, data is not changed, but only labeled with its cluster id.


Let's now focus on the **Shifting Step**, in this example. For the simply matter of simplicity, let‚Äôs use a **single data point** as an example to show how the shifting processes. 

Suppose we have a point located at the edge of the bottom right cluster (black dot shown in the plot below).
<p style="text-align:right;"><img width='40%' src="figs/mean5.png" />
    
After calculating the mean value of feature_1 and that of feature_2 within the local area defined by the bandwidth parameter, we get the **location of the mean point** (shown as the red dot above). Then the **center of circle moves from the black dot to the red one**.

Next, the aforementioned procedure is repeated at the new point position as shown below.
<p style="text-align:right;"><img width='40%' src="figs/mean6.png" />


After sufficient time of iterations we will go towards the situation in which the number of points within the black circle no more increase (we are converging). Therefore, the **original data point arrives at its destination**, the centroid of the cluster it belongs to.
<p style="text-align:right;"><img width='40%' src="figs/mean7.png" />


##### VERY IMPORTANT REMARK: 
You may have noticed that the **calculation will be super redundant** if this procedure is performed for every data point. So, in practice, **when these moving circles overlap**, only the **one containing the most points is preserved**.

Since **Mean Shift Clustering** only has one parameter, the **Bandwidth** (or the window size in some contexts), it needs super caution when choosing a appropriate value for it.

#### a.1.7.6: MEAN-SHIFT CLUSTERING: PROs & CONs

Looking at the **Mean-Shift Clustering Advantages**, we find that:
- It is a **very simple** Clustering Algorithm that works very well on **Spherical-Shaped** data. 
- It **automatically selects the number of clusters** contrary to other clustering algorithms like KMeans. 
- The output of Mean Shift is not dependent on the initialization since, at the start, each point is a cluster.

Looking at the **Mean-Shift Clustering Disadvantages**, we find that:
- It is **not highly scalable**, as it requires multiple nearest neighbor searches during its execution. 
- It has a **quite high Time Complexity** of O(n^2). 
- Manually choosing the bandwidth can be non-trivial, and selecting a wrong value can lead to bad results (clearly if we decide to not use the Sklearn's built-in function).
- It also fails to find the correct clusters for some data-sets. 


#### a.1.7.7: MEAN-SHIFT CLUSTERING: COMPARISON VS K-MEANS

Let's now compare the **Mean-Shift Algorithm** with the **K-Means**. One of the most important difference is that K-means makes two broad assumptions: 
- The number of clusters is already known.
- The clusters are shaped spherically (or elliptically). 

Mean shift , being a non parametric algorithm, does not assume anything about number of clusters. The number of modes give the number of clusters. Also, since it is based on density estimation, it can handle arbitrarily shaped clusters.

Furthermore, **K-means** is very **sensitive to initializations**. A wrong initialization can delay convergence or some times even result in wrong clusters. **Mean Shift** is fairly robust to initializations. Typically, mean shift is run for each point or some times points are selected uniformly from the feature space. Similarly, K-means is sensitive to outliers but Mean Shift is not very sensitive.

**K-Means** is fast and has a time complexity O(knT) where k is the number of clusters, n is the number of points and T is the number of iterations. Classic **Mean Shift** is computationally expensive with a time complexity O(Tn^{2}).
Mean shift, on the other hand, is sensitive to the selection of bandwidth, h.

#### a.1.8.1: DBSCAN: IDEA

The main idea behind **DBSCAN** is that a point belongs to a cluster if it is close to many points from that cluster.

There are two key **DBSCAN Parameters**:
- $\epsilon (\text{eps})$: It is defined as the the **distance** that specifies the **neighborhoods**. It can be compared to the **Bandwidth Parameter** in **Mean-Shift Clustering**. We say that two points $p1$ and $p2$ are considered to be neighbors if: $d(p1, p2) \leq \epsilon$.
- $\text{minPts}$: it is defined as the **Minimum Number of data points to define a cluster**.

<img src="https://dashee87.github.io/images/DBSCAN_tutorial.gif" width="500" align="center">



#### a.1.8.2: DBSCAN: POINT CLASSIFICATION
Depending the previous two parameters, we can **Classify Data Points** in: 
- **Core Point**: A point $p_{core}$ is defined to be a **Core Point** if there are at least $\text{minPts}$ number of points (including the point itself) in its surrounding area with radius $\epsilon$.
    $$(\sum_{i = 1}^n \delta_i )\geq minPts \quad \quad \text{where }
\delta_{i} = \begin{cases}
1 &\text{if } d(p_i, p_{core}) \leq \epsilon  \\
0 &\text{if } d(p_i, p_{core}) > \epsilon   \end{cases}
$$
- **Border Point**: A point $p_{bord}$ is defined to be a **Border Point** if:
    - It is reachable from a core point $p_{core}$. 

    - There are less than minPts number of points within its surrounding area (It is not a **Core Point**).
    $$(\sum_{i = 1}^n \delta_i ) < \text{minPts} \quad \quad \text{where }
\delta_{i} = \begin{cases}
1 &\text{if } d(p_i, p_{bord}) \leq \epsilon  \\
0 &\text{if } d(p_i, p_{bord}) > \epsilon   \end{cases}
$$
- **Outlier**: A point $p_{out}$ is defined to be an **Outlier** if: 
    - It is not a Core Point. $$(\sum_{i = 1}^n \delta_i ) < \text{minPts} \quad \quad \text{where }
\delta_{i} = \begin{cases}
1 &\text{if } d(p_i, p_{bord}) \leq \epsilon  \\
0 &\text{if } d(p_i, p_{bord}) > \epsilon   \end{cases}
$$
    - It is not reachable from any core points.

Let's make an example for better understanding this distinction. Let's consider $\text{minPts} = 4$. 
<p style="text-align:left;"><img width='30%' src="figs/scan5.png" />

What we can see from the example is that: 
- **Red Points** are **Core Points** because there are at least 4 points within their surrounding area with radius $\epsilon$. This area is shown with the circles in the figure. 
- **Yellow Points** are **Border Points** because they are reachable (being in the surrounding area of a core point) from a core point and have less than 4 points within their neighborhood. The points B and C have two points (including the point itself) within their neigborhood (i.e. the surrounding area with a radius of eps). 
- **N** is an **Outlier** because it is not a core point and cannot be reached from a core point.

#### a.1.8.3: DBSCAN: ALGORITHM

The **DBSCAN Algorithm** can be subdivided in the folowing steps:
- **Step 1:** $\text{minPts}$ and $\epsilon$ are determined.
- **Step 2:** A **Starting Point** is selected at random and it‚Äôs neighborhood area is determined using the $\epsilon$ radius parameter. If there are at least $\text{minPts}$ number of points in the neighborhood, the point is marked as **Core Point** and a **Cluster Formation Starts**. If not, the point is marked as **Noise**(it may be revisited by other points and may be part of a cluster). Once a cluster formation starts (let‚Äôs call it cluster A), all the points within the neighborhood of initial point become a part of cluster A. If these new points are also core points, the points that are in the neighborhood of them are also added to cluster A. (Think at this step as a sort of recursive step).
- **Step 3**: Randomly **choose another point** among the points that have not been visited in the previous steps. Then same procedure applies. This process is finished when all points are visited. 

Note that, as **Distance Meansurment** it is usually used the **Euclidean Distance**.

The required condition **to form a cluster** is to **have at least one core point**. Although very unlikely, we may have a cluster with only one core point and its border points.



#### a.1.8.4: DBSCAN: PARAMETER ESTIMATION

We know that the choice for the parameters is very important and does change the result of the **DBSCAN Application**. 

- **minPts**: For what concerns the first parameter, the minimal number of points that must be present in order to define a cluster, it is commonly used a sort of **rule of Thumb**, that can be derived from the number of dimensions D in the data set, and it is defined to be **minPts ‚â• D + 1**. Note also that since the choice minPts = 1 does not make sense, as then every point on its own will already be a cluster, and since the fact that the choice minPts ‚â§ 2 will result in having the same result of a hierarchical clustering, **minPts must be chosen at least = 3**. However, **larger values** are usually **better for data sets with noise** and will yield more significant clusters. As a rule of thumb, **minPts = 2¬∑D** can be used, but it may be necessary to choose larger values for very large data, for noisy data or for data that contains many duplicates. 

- $\epsilon$: For what concerns the other Parameter, we have that obtaining an estimation is quite less straightforward. One of the main technique used is described in the paper lnked below and has it is known as the **Nearest Neighbours Elbow**. This technique calculates the **average distance between each point and its k nearest neighbors**, where **k = the MinPts** value you selected. The average k-distances are then plotted in ascending order on a k-distance graph. You‚Äôll find the optimal value for Œµ at the **point of maximum curvature**. In order to get it automatically, I can use the same tool used for the Elbow Method: the **Knee Locator**.

https://iopscience.iop.org/article/10.1088/1755-1315/31/1/012012/pdf


####  a.1.8.5: DBSCAN: PROs & CONs

We can say that the **DBSCAN Advatnages** are:
- It does not require to specify number of clusters beforehand.
- It performs well with arbitrary shapes clusters.
- It is robust to outliers and able to **detect the outliers**.

However, it has also some **Disadvantages**, that are:
- It is not easy, in some cases, to determine an appropriate distance of neighborhood $\epsilon$ and it requires domain knowledge.
- It suffers from the **Curse of Dimensionality**. This means that distance functions become less meaningful in higher dimensions, as all points are ‚Äòfar away‚Äô from each other.

Let's now see a **Practical Example**, in such a way to see all about the **Nearest Neighbour Elbow**, the **DBSCAN** at work, and one of its main application: the **Noise Detection**.

Let's start with creating a sample dataset that will be Semi-Moon like shaped. After creating the Dataset, let's apply the **DBSCAN** Clustering in such a way to see how it behaves.  

We will then choose minPts = 4, since the Dimension is D = 2. Therefore, using the rule of Thumb, we obtain $minPts = 2*D = 4$. For the $\epsilon$, we choose it to be equal to $\epsilon = 0.1$, literally a random number. 

What we can observe, is that it clusters pretty well data, but it leaves out lots of outliers, and, it gets a wrong number of clusters. 

```python
#SETTING THE SEED:
np.random.seed(42)                                                      #Setting up the seed for reproducibility. 

#CREATION OF THE DATASET:
X, y = make_moons(n_samples=500, noise=0.1)                             #Creation of the Dataset moon-like shaped.
fd=pd.DataFrame(X,y)                                                    #Making it to be a Pandas Dataframe.
fd=fd.rename(columns={0: "X1", 1:"X2"})                                 #Rename the Dataframe's Columns.

#CREATION OF A DBSCAN RANDOM CLUSTERING:
dbscan_cluster = DBSCAN(eps=0.1, min_samples=4).fit(X)                  #Computing the DBSCAN Clustering.

#NUMBER OF CLUSTERS ESTIMATION:
labels=dbscan_cluster.labels_                                           #Taking the labels from the DBSCAN Clustering.
N_clus=len(set(labels))-(1 if -1 in labels else 0)                      #Computing the number of Estimated Clusters I have found.
print('Estimated no. of clusters: %d' % N_clus)                         #Printing out the number of Estimated Clusters I have found.

#IDENTIFICATION OF NOISY POINTS:
n_noise = list(dbscan_cluster.labels_).count(-1)                        #Computing the number of Estimated Noisy Points.
print('Estimated no. of noise points: %d' % n_noise)                    #Printing out the number of Estimated Noisy Points I have found.

#DBSCAN CLUSTERING VISUALIZATION:
plt.scatter(X[:, 0], X[:, 1], c=dbscan_cluster.labels_, label=y, cmap='inferno')#Plotting the Data coloured depending on the DBSCAN's label.
plt.xlabel('Feature $X1$')                                              #Inserting the x-axis label.
plt.ylabel('Feature $X2$')                                              #Inserting the y-axis label.
plt.title('DBSCAN Clustering Resolution')                               #Inserting the title in the plot.
plt.show()                                                              #Showing the result obtained.
```

Since I want to obtain the best result I can, I proceed as follows. I have taken the best choice I could for the $\text{minPts} = 4$. Now, I must get the best result for what concerns the other parameter: the $\epsilon$. 

In order to this, I need to apply the **Nearest Neighbours Elbow**. Here, indeed, I will have to find the **point of maximum curvature**, that will be my best $\epsilon$ 's value. Let's apply this tool and see what it comes out.

From the ****Nearest Neighbours Elbow**** Method, the **Knee Locator** individuates, as **point of maximum curvature**, to be equal to about 0.13. I therefore make the other parameter choice for which: $\epsilon = 0.139$.

```python

#FINDING THE BEST VALUE FOR EPSILON USING THE NEAREST NEIGHBOUR ELBOW:
nearest_neighbors = NearestNeighbors(n_neighbors=8)                     #Computing the Nearest-Neighbours with 4 = 2D.
neighbors = nearest_neighbors.fit(fd)                                   #Fitting the Dataset.
distances, indices = neighbors.kneighbors(fd)                           #Taking Distances and Indices.
distances = np.sort(distances[:,7], axis=0)                             #Sorting the distances in ascending order.

#BUILDING UP A KNEE LOCATOR:
i = np.arange(len(distances))                                           #Creation of a range of length distances.
knee = KneeLocator(i, distances, S=1, curve='convex',                   #Building a Knee Locator that will find out what is the best value for epsilon.
                   direction='increasing', interp_method='polynomial')

#VISUALIZING THE NEAREST NEIGHBOUR METHOD RESULT:
fig = plt.figure(figsize=(5, 5))                                        #Creatio of the Figure.
knee.plot_knee()                                                        #Plotting the Knee locator.
plt.plot(distances)                                                     #Plotting the distances sorted.
plt.xlabel('Points')                                                    #Inserting the x-axis label.
plt.ylabel('Distance')                                                  #Inserting the y-axis label.
plt.title('Nearest Neighbour Method')                                   #Inserting the title in the plot.
plt.show()                                                              #Showing the result obtained.

#PRINTING OUT THE BEST VALUE:
print("The best Value for Epsilon is: ", distances[knee.knee])          #Printing out the best value for Epsilon.
```

After having individuated also the best value for $\epsilon$, I can recompute again the DBSCAN with the new optimal values obtained from the studies:
- $minPts = 2*D = 4$
- $\epsilon = 0.139$

What I got now, is an almost perfect Clustering, for which I have that 2 clusters have beenn individuated, and only 5 outliers are detected. 

```python
#CREATION OF A DBSCAN RANDOM CLUSTERING:
dbscan_cluster = DBSCAN(eps=0.139, min_samples=4).fit(X)                #Computing the DBSCAN Clustering.

#NUMBER OF CLUSTERS ESTIMATION:
labels=dbscan_cluster.labels_                                           #Taking the labels from the DBSCAN Clustering.
N_clus=len(set(labels))-(1 if -1 in labels else 0)                      #Computing the number of Estimated Clusters I have found.
print('Estimated no. of clusters: %d' % N_clus)                         #Printing out the number of Estimated Clusters I have found.

#IDENTIFICATION OF NOISY POINTS:
n_noise = list(dbscan_cluster.labels_).count(-1)                        #Computing the number of Estimated Noisy Points.
print('Estimated no. of noise points: %d' % n_noise)                    #Printing out the number of Estimated Noisy Points I have found.

#DBSCAN CLUSTERING VISUALIZATION:
plt.scatter(X[:, 0], X[:, 1], c=dbscan_cluster.labels_, label=y, cmap='inferno')#Plotting the Data coloured depending on the DBSCAN's label.
plt.xlabel('Feature $X1$')                                              #Inserting the x-axis label.
plt.ylabel('Feature $X2$')                                              #Inserting the y-axis label.
plt.title('DBSCAN Clustering Resolution Optimized')                     #Inserting the title in the plot.
plt.show()                                                              #Showing the result obtained.
```

The last step now that I am going to apply, is to demonstate the main **DBSCAN Application**. What I am speaking about, is the **Noise Detection** Advantage that **DBSCAN** has. Thanks to this feature, the **DBSCAN** can also be followed by a **Noise Removal Process**, in which all the **Noise** that has been detected, it is removed. (It can be compared to a sort of **Cook's Distance** or also to the **Leverage**, used both to eliminate **Highly Influential Points** in the **Linear Regression Process**).

After applying the **Noise Removal Process**, I can apply again **DBSCAN**. What I can observe is that now, 0 Outliers have been detected, and the Clustering is perfect. 

```python
#NOISE REMOVAL:
range_max = len(X)                                                              #Computing the length of the Dataset.
X = np.array([X[i] for i in range(0, range_max) if labels[i] != -1])            #Taking the new Dataset, that does not contain any noise point.
labels = np.array([labels[i] for i in range(0, range_max) if labels[i] != -1])  #Taking all the labels of the new Denoised Dataset.

#CREATION OF A DBSCAN RANDOM CLUSTERING:
dbscan_cluster = DBSCAN(eps=0.139, min_samples=4).fit(X)                        #Computing the DBSCAN Clustering.

#NUMBER OF CLUSTERS ESTIMATION:
labels=dbscan_cluster.labels_                                                   #Taking the labels from the DBSCAN Clustering.
N_clus=len(set(labels))-(1 if -1 in labels else 0)                              #Computing the number of Estimated Clusters I have found.
print('Estimated no. of clusters: %d' % N_clus)                                 #Printing out the number of Estimated Clusters I have found.

#IDENTIFICATION OF NOISY POINTS:
n_noise = list(dbscan_cluster.labels_).count(-1)                                #Computing the number of Estimated Noisy Points.
print('Estimated no. of noise points: %d' % n_noise)                            #Printing out the number of Estimated Noisy Points I have found.

#DBSCAN CLUSTERING VISUALIZATION:
plt.scatter(X[:, 0], X[:, 1], c=dbscan_cluster.labels_, label=y, cmap='inferno')#Plotting the Data coloured depending on the DBSCAN's label.
plt.xlabel('Feature $X1$')                                                      #Inserting the x-axis label.
plt.ylabel('Feature $X2$')                                                      #Inserting the y-axis label.
plt.title('DBSCAN Clustering Resolution Noise Removal')                         #Inserting the title in the plot.
plt.show()                                                                      #Showing the result obtained.
```

### a.1.9: GAUSSIAN MIXTURE MODEL (GMM) CLUSTERING

Let's breifly go through the already explained ****GMM Clustering**, faced during the Professor's Lectures. We know that **Gaussian Mixture Models (GMMs)** assume that there are a certain number of Gaussian distributions, and each of these distributions represent a cluster. Hence, a Gaussian Mixture Model tends to group the data points belonging to a single distribution together.

Gaussian Mixture Models are **Probabilistic Models** and use the so called **Soft Clustering** approach (w.r.t. the K-Means Hard Clustering) for distributing the points in different clusters. 

The **Gaussian Mixture Model** is the Computer Scientist way of calling this method. This method, has its **density function** that is a linear combination of Gaussian distributions, which is called **Mxture of Gaussians (MoG)**.

$$p(\mathbf{x})=\sum_{k=1}^{K} \pi_{k} \mathcal{N}\left(\mathbf{x} \mid \mu_{k}, \Sigma_{k}\right)$$

where $\pi_{k}$ the **mixing coefficients**, where:
$$
\sum_{k=1}^{K} \pi_{k}=1 \quad \text { and } \quad \pi_{k} \geq 0 \quad \forall k
$$

It requweres two parameters, the **Mean** and the **Covariance**, to describe the position and shape of each cluster.

The model is based on the **Multivariate Gaussian Distribution** which is: $$ N(x: \mu, \Sigma) = \frac{1}{(2\pi\Sigma)^{1/2}} \exp(-\frac{1}{2} (x-\mu)^{T}\Sigma^{-1}(x-\mu))$$

where for each cluster: 
- **x :** Vector of a data point.
- **$\mu$:** Vector of the Mean of each Cluster.
- **$\Sigma$:** Covariance Matrix between each Dimension.

#### a.1.9.1: GMM CLUSTERING: EXPECTATION MAXIMIZATION(EM) ALGORITHM

**Expectation-Maximization (EM)** is a statistical algorithm for finding the right model parameters. We typically use EM when the data has missing values, or in other words, when the data is incomplete.

These **missing variables** are called **Latent Variables**. We consider the target (or cluster number) to be unknown when we‚Äôre working on an unsupervised learning problem.

Since we do not have the values for the latent variables, Expectation-Maximization tries to use the existing data to determine the optimum values for these variables and then finds the model parameters. Based on these model parameters, we go back and update the values for the latent variable, and so on.

The Algorithm corresponds in the following steps:
- **Initialization Step**: Initialize the means $\mu_{k}$, covariances $\Sigma_{k}$ and mixing coefficients $\pi_{k}$ parameters. 
- Iterate until convergence through the following two steps:
    - **E-STEP**: In this step, the available data is used to estimate (guess) the values of the missing variables (**Latent Variables**). For each point $x_i$, calculate the probability that it belongs to cluster/distribution $c_1, c_2, ‚Ä¶ c_k$. This is done using the below formula:
$$\gamma_{k} =\frac{\text{probability}\space x_i \in \text{c}}{\text{sum of probability}\space x_i \in c_1, c_2, ..., c_k } \quad = \quad \frac{
 \mathcal{N}(x_i;\mu_c,\sigma_c)\cdot\pi_c}
{\sum_{c'}\mathcal{N}(x_i;\mu_{c'},\sigma_{c'})\cdot\pi_{c'}}
$$
This value will be high when the point is assigned to the right cluster and lower otherwise.

- **M-STEP**: Based on the estimated values generated in the E-step, the complete data is used to update the parameters. In this step, we go back and update $\pi$, $\mu$ and $\sigma$ values. These are updated in the following manner:
    - $\pi_k$: The new density is defined by the ratio of the number of points in the cluster and the total number of points: $$\pi_{k}=\frac{N_{k}}{N} \text { with } N_{k}=\sum_{n=1}^{N} \gamma_{k}^{(n)}$$
    - $\mu_k \text{and} \sigma_k$: The mean and the covariance matrix are updated based on the values assigned to the distribution, in proportion with the probability values for the data point. Hence, a data point that has a higher probability of being a part of that distribution will contribute a larger portion:
    $$\mu_{k}=\frac{1}{N_{k}} \sum_{n=1}^{N} \gamma_{k}^{(n)} \mathbf{x}^{(n)} $$
    $$\Sigma_{k}=\frac{1}{N_{k}} \sum_{n=1}^{N} \gamma_{k}^{(n)}\left(\mathbf{x}^{(n)}-\mu_{k}\right)\left(\mathbf{x}^{(n)}-\mu_{k}\right)^{T} $$
    
- Evaluate log likelihood and check for convergence
$$
\ln p(\mathbf{X} \mid \pi, \mu, \Sigma)=\sum_{n=1}^{N} \ln \left(\sum_{k=1}^{K} \pi_{k} \mathcal{N}\left(\mathbf{x}^{(n)} \mid \mu_{k}, \Sigma_{k}\right)\right)
$$

#### INITIALIZATION STEP:

Note that, for the **Initialization Step**, there are several choices, as in the K-Means case:
- **Random Initialization**: Centers are chosen as a small perturbation away from the mean of all data. This method is simple but can lead to the model taking longer to converge.
- **K-Means++**: Respondibilities are initialized using the K-Means++ Algorithm. This uses the initialization method of k-means clustering: k-means++. This will pick the first center at random from the data. Subsequent centers will be chosen from a weighted distribution of the data favouring points further away from existing centers. k-means++ is the default initialization for k-means so will be quicker than running a full k-means but can still take a significant amount of time for large data sets with many components.

####  a.1.9.2: GMM CLUSTERING: HOW MANY GAUSSIANS? 

As in the case of the **K-Means**, there is the problem also for **GMM** of determining what is the **Optimal Number of Components (Gaussians)** for a given dataset.

We have different options:
- **Cross-Validation**: A generative model is inherently a probability distribution for the dataset, and so we can simply evaluate the likelihood of the data under the model, using cross-validation to avoid over-fitting. 
- **Information Criterion**: Another means of correcting for over-fitting is to adjust the model likelihoods using some analytic criterion such as the **Akaike Information Criterion (AIC)** or the **Bayesian Information Criterion (BIC).** $$ \text{(AIC)}  \qquad \arg\min_K \mathcal{L}_K  + \underbrace{K\cdot 2D}_{\text{regularization}}$$
$$ \text{(BIC)} \qquad \arg\min_K \mathcal{L}_K  + \underbrace{K\cdot\log(D)}_{\text{regularization}}$$

### b.1.0: SILHOUETTE SCORE: CLUSTERING ALGORITHM COMPARISON

The first analysis I will perform is about the **Silhoette Score**. I have already explained it during the **Choice of the best K** section, but in that case, it was for choosing what was the best number of clusters. 

In this case, the **Silhoette Score** will help me in choosing what is the best algorithm under this point of view. From what we knew from the choice of the best K section, we have that the **Silhouette Coefficient** is a measure of **Cluster Cohesion and Separation**, used to evaluate the **Quality of Clusters** in terms of how well samples are clustered with other samples that are similar to each other. It is used to measure **how dense and well-separated the clusters are**. 

We also know that the **Silhouette Coefficient ranges between -1 and 1**. The larger the Silhouette Coefficient, the closer samples are to their cluster w.r.t. they are to other clusters. We have that: 
- If the score is 1, the cluster is dense and well-separated than other clusters. 
- A value near 0 represents overlapping clusters with samples very close to the decision boundary of the neighbouring clusters.
- A negative score [-1, 0] indicate that the samples might have got assigned to the wrong clusters.

The **Silhouette Coefficient** is calculated for each sample of different clusters. In order to calculate the Silhouette score for each observation / data point, the following distances need to be found out for each observations belonging to all the clusters:

‚Ä¢ **Mean Intra-Cluster Distance:** Mean distance between the observation and all other data points in the same cluster. It is denoted by ***a***.

‚Ä¢ **Mean Nearest-Cluster Distance:** Mean distance between the observation and all other data points of the next nearest cluster. It is denoted by ***b***.

The **Silhouette Coefficient S**, for each sample is calculated using the following formula:

$$S = \frac{(b-a)}{max(a,b)} = \frac{Mean Nearest Cluster Distance - Mean Intra Cluster Distance}{max(Mean Intra Cluster Distance, Mean Nearest Cluster Distance)}$$

I am now going to apply this tool over the 15 versions of algorithm I have introduced in the first macro-section of the Project. 

##### GOAL: 
My aim is to find what is the **Best Algorithm** taking into account the **Silhouette Score**.

##### RESULT:
As a result of this first analysis, I can observe that all the Algorithms (**K-Means Versions, K-Means++ Versions, Naive-Sharding Versions, K-Medoids and K-Medians Versions**) have the same $\text{Silhouette Score} = 0.90082$, that is therefore the best result.

The only two other algorithms, have a lower score, that is $\text{Silhouette Score} = 0.76804$ and $\text{Silhouette Score} = 0.84353$ for **GMM** with **Random Initialization** and **GMM** with **K-Means++ Initialization** respectively.

Since we know that the more the **Silhoette Score** is towards 1 and the better it is, the other algorithms are better thatn the GMM versions, under sthis point of view.

### b.1.3: PURITY

Let $\text{N}_{ij}$ be the number of objects in cluster $i$ that belong to class $j$, and let $N_i = \sum_{j=1}^C N_{ij}$ be the total number of objects in cluster i. 

Define $p_{ij} = \frac{N_{ij}}{N_i}$ to be the **Empirical Distribution** over class
labels for cluster i. 

We define the **Purity** of a Cluster as $pi = max_j \space p_{ij}$ , and the **Overall Purity** of a
**Clustering** as: $$\text{purity} = \sum_{i} \frac{N_i}{N} \space p_i$$

As an example, let's consider the following Clustering:
<p style="text-align:left;"><img width='30%' src="figs/measure1.png" />
    
We have that its purity is:
$$\text{purity} = \sum_{i} \frac{N_i}{N} \space p_i = \frac{6}{17}\frac{5}{6}\space + \space \frac{6}{17}\frac{4}{6} \space + \space \frac{5}{17}\frac{3}{5} = 0.71 $$


What we have is that the **Purity ranges between 0 (bad) and 1 (good)**. 

However, we can trivially achieve a **Purity = 1** by **putting each object into its own cluster**. 

So, we have that, this measure does not penalize for the number of clusters.

### b.1.4: RAND INDEX

Let's define $U = {u_1, ... , u_R}$ and $V = {v_1, ... , v_C}$ be two different partitions of the $N$ data points. For example, U might be the estimated clustering and V is reference clustering derived from the class labels.

Now, let's define a **2 x 2 Contingency Table**, containing the following numbers: 
- **TP (True Positive, a.k.a. "a"):**  It is the number of pairs that are in the same cluster in both U and V (true positives). 
- **FN (False Negative, a.k.a. "b"):** It is the number of pairs that are in the different clusters in U but the same cluster in V (false negatives). 
- **FP (False Positive, a.k.a. "d"):** It is the number of pairs that are in the same cluster in U but different clusters in V (false positives).
- **TN (True Negative, a.k.a. "d"):** It is the number of pairs that are in the different clusters in both U and V (true negatives).

||**Pairs Assigned to the Same Cluster U**  |  **Pairs Assigned to a Different Cluster U**  | 
| --- | --- | --- |
| **Pairs Assigned to the Same Cluster V** | TP | FN |
| **Pairs Assigned to a Different Cluster V** | FP | TN |


We can define the **Rand Index** as: $$R = \frac{TP + TN}{TP + FP + FN + TN}$$

This can be interpreted as the fraction of clustering decisions that are correct. Note that, clearly, we have that $0 \leq R \leq 1$.

#### b.1.4.1: RAND INDEX: PROs & CONs 

Let's see the advantages and disadvantages of the **Rand Index**:

The **Rand Index Advantages are**:
- It is proportional to the number of sample pairs whose labels are the same in both labels_pred and labels_true, or are different in both.
- Lower values indicate different labelings. Similar clusterings have a high Rand index. We have that 1.0 is the perfect match score. The score range is [0, 1]. 
- No assumption is made on the cluster structure. We can use the Rand Index to compare all kinds of clustering algorithms, and can be used to compare clustering algorithms.

The **Rand Index Disadvantages are**:
- The **Rand Index** requires knowledge of the ground truth classes which is almost never available in practice.
- The **Rand Index** is often close to 1.0 even if the clusterings themselves differ significantly. This can be understood when interpreting the Rand index as the accuracy of element pair labeling resulting from the clusterings: In practice there often is a majority of element pairs that are assigned the different pair label under both the predicted and the ground truth clustering resulting in a high proportion of pair labels that agree, which leads subsequently to a high score.

#### b.1.4.2:  ADJUSTED RAND INDEX

The **Rand Index** does not guarantee that random label assignments will get a value close to zero (esp. if the number of clusters is in the same order of magnitude as the number of samples).

To counter this effect we can discount the expected RI, $E(R)$ of random labelings by defining the **Adjusted Rand Index** as follows:
$$AdjR = \frac{R - E(R)}{\text{max}(R) - E(R)}$$

For the **Adjusted Rand Index**, we have that the range is changed to [-1, 1].

#### EXAMPLE:
Let's make an example. Let's say we have a dataset of 5 items and we have the following 2 Clustering Assignments:
- Clustering U: 1,2,2,1,1
- Clustering V: 2,1,2,1,1

The two different clustering assignment consist of the same set of n points, i.e. n=5. 

In order to calculate the **Rand Index**, we have to compare pairs. In total, we have that there are: $TP + FP + FN + TN = {n \choose 2}$ possible pairs of n points. In our example, we have 5 points in each clusterings, so we have: $$TP + FP + FN + TN = {5 \choose 2} = 10$$

In order to calculate $TP + FP + FN + TN$, we need to first list the possible pairs in terms of indices that correspond to the elements of the two clusterings. Let's set them all to be = 0. (TP = 0, FP = 0,  FN = 0, TN = 0).

The possible pairs are; $$[1,2],[1,3],[1,4],[1,5],[2,3],[2,4],[2,5],[3,4],[3,5],[4,5]$$

These, are the pairs corresponding to the $[i,j]$ i-th element and j-th element in the two clusters.

For instance, $[1,2]$ corrresponds to $[1,2]$ in Cluster U, and to $[2,1]$ in Cluster V.

So, let's start the computations:
- $[1,2]$: We are now going to see the first and second element for both the U and V Cluster:
    - Cluster U: $[1,2] \rightarrow 1 \neq 2 $
    - Cluster V: $[2,1] \rightarrow 2 \neq 1 $
    - RESULT: $TN = TN + 1$
- $[1,3]$: We are now going to see the first and third element for both the U and V Cluster:
    - Cluster U: $[1,2] \rightarrow 1 \neq 2 $
    - Cluster V: $[2,2] \rightarrow 2 = 2 $
    - RESULT: $FN = FN + 1$
- $[1,4]$: We are now going to see the first and fourth element for both the U and V Cluster:
    - Cluster U: $[1,1] \rightarrow 1 = 2 $
    - Cluster V: $[2,1] \rightarrow 2 \neq 1 $
    - RESULT: $FP = FP + 1$

- $[1,5]$: We are now going to see the first and fifth element for both the U and V Cluster:
    - Cluster U: $[1,1] \rightarrow 1 = 2 $
    - Cluster V: $[2,1] \rightarrow 2 \neq 1 $
    - RESULT: $FP = FP + 1$
- $[2,3]$: We are now going to see the second and third element for both the U and V Cluster:
    - Cluster U: $[2,2] \rightarrow 1 = 2 $
    - Cluster V: $[1,2] \rightarrow 2 \neq 2 $
    - RESULT: $FP = FP + 1$
- $[2,4]$: We are now going to see the second and fourth element for both the U and V Cluster:
    - Cluster U: $[2,1] \rightarrow 2 \neq 1 $
    - Cluster V: $[1,1] \rightarrow 1 = 1 $
    - RESULT: $FN = FN + 1$

- $[2,5]$: We are now going to see the second and fifth element for both the U and V Cluster:
    - Cluster U: $[2,1] \rightarrow 2 \neq 1 $
    - Cluster V: $[1,1] \rightarrow 1 = 1 $
    - RESULT: $FN = FN + 1$
- $[3,4]$: We are now going to see the third and fourth element for both the U and V Cluster:
    - Cluster U: $[2,1] \rightarrow 2 \neq 1 $
    - Cluster V: $[2,1] \rightarrow 2 \neq 1 $
    - RESULT: $TN = TN + 1$
- $[3,5]$: We are now going to see the third and fifth element for both the U and V Cluster:
    - Cluster U: $[2,1] \rightarrow 2 \neq 1 $
    - Cluster V: $[2,1] \rightarrow 2 \neq 1 $
    - RESULT: $TN = TN + 1$

- $[4,5]$: We are now going to see the fourth and fifth element for both the U and V Cluster:
    - Cluster U: $[1,1] \rightarrow 1 = 1 $
    - Cluster V: $[1,1] \rightarrow 1 = 1 $
    - RESULT: $TP = TP + 1$


At the very end we get that: 

||**Pairs Assigned to the Same Cluster U**  |  **Pairs Assigned to a Different Cluster U**  | 
| --- | --- | --- |
| **Pairs Assigned to the Same Cluster V** | 1 | 3 |
| **Pairs Assigned to a Different Cluster V** | 3 | 3 |

Therefore, the final **Rand Index** is equal to:
$$R = \frac{TP + TN}{TP + FP + FN + TN} = \frac{1 + 3}{1 + 3 + 3 + 3} = 0.4$$

### b.1.5: DAVIES-BOULDIN INDEX

As I mentioned earlier, for my special example **Idle VS Working**, I have the luck to have the truth labels, given the fact that I built the Dataset by myself. However, again, in a clear unsupervised setting, in most of the cases, we do not have this privileges. 

For this reason, I will now also opt for an analysis for which I behave like if I hadn't any truth-labels. (As a small remark, note that we have already seen one **Performance Measure** of this type, the **Silhouette Coefficient**

If the ground truth labels are not known, the **Davies-Bouldin Index** could be one option. It can be used to evaluate the model, where a **lower Davies-Bouldin Index** relates to a model with **better separation** between the clusters.
This index signifies the **average ‚ÄòSimilarity‚Äô** between clusters, where the similarity is a measure that compares the distance between clusters with the size of the clusters themselves.
Zero is the lowest possible score. Values closer to zero indicate a better partition.


The index is defined as the average similarity between each cluster $C_i$ for $i = 1, ... , k$  and its most similar one $C_j$. 

In the context of this index, similarity is defined as a measure  that trades off:
- $s_i$: It is defined as the average distance between each point of cluster $i$  and the centroid of that cluster (a.k.a **Cluster Diameter**).
- $d_{ij}$: It is defined as the distance between cluster centroids $i$ and $j$.

A simple choice to construct $R_{ij}$ so that it is nonnegative and symmetric is: $$R_{ij} = \frac{s_i + s_j}{d_{ij}}$$

Then, the **Davies-Bouldin Index** is defined as:
$$DB = \frac{1}{k}\sum_{i=1}^k max_{i \neq j} R_{ij}$$

#### b.1.5.1: DAVIS-BOULDIN INDEX: COMPUTATION STEPS

In order to compute the **Davis-Bouldin Index** let's divide the procedure in 4 steps:
- **Step 1: (Calculate Intra-Cluster Dispersion)** Let's define the Cluster i **Dispersion** as: $$\sqrt[q]{{\frac{1}{T_i}\sum_{j=1}^{T_i} |X_j - A_j|^q}}$$
where we have that:
    - $i$ : It defines a particular identified cluster.
    - $T_i$ It defines the number of vectors (observations) in cluster $ùëñ$.
    - $X_j$ : It defines the j-th vector (observation) in cluster $i$.
    - $A_i$ : It defines the centroid of cluster $i$.
    - $q$ : It is usually set to 2, which calculates the **Euclidean Distance**. 
    
Basically, to get the **Intra-Cluster Dispersion**, we calculate the average distance between each observation within the cluster and its centroid. 

Let's assume that through one of the Clustering Technique was used on some dataset in order to generate three clusters. 

Using the above formula, the intra-cluster dispersion would be calculated for each of the clusters and we will have values for $S_1$, $S_2$, and $S_3$.

<p style="text-align:left;"><img width='50%' src="figs/dbindex1.png" />


- **Step 2: (Calculate Separation Measure)** Let's define the **Separation between two clusters $i$ and $j$ as:** $$M_{ij = }\sqrt[p]{{\sum_{k=1}^{N} |a_{ki} - a_{kj}|^p}} = ||A_i -A_j ||_p$$
where we have that:
    - $a_{ki}$ : It defines the $\text{k-th}$ component of n-dimensional centroid $A_i$. 
    - $a_{kj}$ : It defines the $\text{k-th}$ component of n-dimensional centroid $A_j$. 
    - $N$ : It defines the total number of clusters.
    - $q$ : It is usually set to 2, which calculates the **Euclidean Distance**. 
    

Let's continue with our example. Using the above formula, we will compute the separation measure for each possible combination of two clusters: $M_{11}, M_{12}, M_{13}, M_{21}, M_{22}, M_{23}, M_{31}, M_{32}, M_{33}$.

Of course, note we also have that: $M_{12} = M_{21}, M_{13} = M_{31}, M_{23} = M_{32}$.

<p style="text-align:left;"><img width='50%' src="figs/dbindex2.png" />


- **Step 3: (Calculate Similarity Between Clusters)** Let's define the **Similarity between two clusters $i$ and $j$ as:** $$R_{ij} = \frac{S_i + S_j}{M_{ij}}$$
where we have that:
    - $S_i$ : It defines the Intra-Cluster Dispersion of Cluster $i$. 
    - $S_j$ : It defines the Intra-Cluster Dispersion of Cluster $j$. 
    - $M_{ij}$ : It defines the distance between centroids of clusters $i$ and $j$.
    
Here we have that the larger $R_{ij}$ is, the more similar clusters $i$ and $j$ are. The best possible case is when these number overall are as low as possible.
    

Similarly to the logic in Step 2, continue to assume that we are working with three clusters. 

Using the above formula, we will compute the similarity for each possible combination of two clusters: $R_{11}, R_{12}, R_{13}, R_{21}, R_{22}, R_{23}, R_{31}, R_{32}, R_{33}$. 

Here, $R_{11} = R_{22} = R_{33}$ are the similarities of each cluster to itself. And we also have $R_{12} = R_{21}, R_{13} = R_{31}, R_{23} = R_{32}$.

- **Step 4: (Find most similar Cluster for each Cluster)** Let's define the **Most Similarity between two clusters $i$ and $j$ as:** $$R_{i} = \text{max}_{i \neq j}(R_{ij})$$
For each cluster $i$ we find the highest ratio from all $R_{ij}$ calculated.

For example, for cluster $i=1$, we calculated $R_{11}, R_{12}, R_{13}$. Continue with $R_{12}$ and $R_{13}$ (since $R_{11}$ doesn‚Äôt satisfy the constraint $i \neq j$, and it makes no sense to compare the cluster with itself).

Now, having $R_{12}$ (which is similarity between clusters 1 and 2) and $R_{13}$ (which is similarity between clusters 1 and 3). From these two measures, we would pick the largest one and refer to the maximum measure as $R_{1}$. Following the same logic we will find $R_{2}$ and $R_{3}$.
    

- **Step 5: (Calculate the Davies-Bouldin Index)** Let's calculate the **Davies-Bouldin Index as:** $$DB = \frac{1}{N}\sum_{i=1}^N R_i$$

which is simply the average of the similarity measures of each cluster with a cluster most similar to it.

Note that the best choice for clusters is where the average similarity is minimized, therefore a smaller DB represents better defined clusters.

#### b.1.5.2: DAVIES-BOULDIN INDEX: PROs & CONs

Let's have a look at **Advantages** and **Disadvantages** of the **Davies-Bouldin Index**. Let's first see its **Advantages**:
- The computation of Davies-Bouldin is simpler than that of Silhouette scores.
- The index is solely based on quantities and features inherent to the dataset as its computation only uses point-wise distances.

Instead, having a look at its **Disadvantages**, we have that:
- The Davies-Bouldin index is generally higher for convex clusters than other concepts of clusters, such as density based clusters like those obtained from DBSCAN.
- The usage of centroid distance limits the distance metric to Euclidean space.

```python
#COMPUTING THE FINAL RESULT:
labels_final = ['K-Means Lloyd', 'K-Means By-Hand', 'K-Means Elkan', 'K-Means Naive Lloyd',     #Labels.
                'K-Means Naive Elkan', 'K-Means++ Lloyd', 'K-Means++ By-Hand', 'K-Means++ Elkan', 
                'K-Medians By-Hand', 'K-Medoids PAM Build', 'K-Medoids Alternate Random',
                'K-Medoids Alternate Heuristic', 'K-Medoids Alternate K-Medoids++',
                'GMM Random', 'GMM K-Means++']

#COMPUTING THE REMAINING ACCURACY SCORE STANDARDIZED:
remaining_accuracies = accuracies = [0.001 - (check_accuracy(klabs_init_lloyd, truth) / 1000),         #Computing the Remaining Accuracies Standardized, namely divided by 1000.
                                        0.001 - (check_accuracy(labs_kmeans_byhand, truth) / 1000), 
                                        0.001 - (check_accuracy(klabs_init_elkan, truth) / 1000),
                                        0.001 - (check_accuracy(klabs_kmeans_naive_lloyd, truth) / 1000),
                                        0.001 - (check_accuracy(klabs_kmeans_naive_elkan, truth) / 1000),
                                        0.001 - (check_accuracy(klabs_kmeans_plus_lloyd, truth) / 1000),                           
                                        0.001 - (check_accuracy(labs_kmeans_plus_byhand, truth) / 1000), 
                                        0.001 - (check_accuracy(klabs_kmeans_plus_elkan, truth) / 1000),
                                        0.001 - (check_accuracy(labs_kmedians_byhand, truth) / 1000),
                                        0.001 - (check_accuracy(kmedoids_pam_manhattan_build.labels_, truth) / 1000),              
                                        0.001 - (check_accuracy(kmedoids_alternate_manhattan_random.labels_, truth) / 1000), 
                                        0.001 - (check_accuracy(kmedoids_alternate_manhattan_heuristic_labels, truth) / 1000),
                                        0.001 - (check_accuracy(kmedoids_alternate_manhattan_k_medoids_plus.labels_, truth) / 1000),
                                        0.001 - (check_accuracy(gmm_label_random, truth) / 1000),
                                        0.001 - (check_accuracy(gmm_label_kmeans_init, truth) / 1000)]

#COMPUTING THE REMAINING SILHOUETTE SCORE STANDARDIZED:
silhouette_scores = [0.001 - (kmeans_random_lloyd_score / 1000),                                #Computing the Remaining Silhouette Score, that will be added to the Final Comparison.
                     0.001 - (kmeans_random_score / 1000), 
                     0.001 - (kmeans_random_elkan_score / 1000), 
                     0.001 - (kmeans_naive_lloyd_score / 1000),
                     0.001 - (kmeans_naive_elkan_score / 1000),
                     0.001 - (kmeans_plus_lloyd_score / 1000),
                     0.001 - (kmeans_plus_byhand_score / 1000),
                     0.001 - (kmeans_plus_elkan_score / 1000),
                     0.001 - (kmedians_byhand_score / 1000),
                     0.001 - (kmedoids_pam_manhattan_build_score / 1000),
                     0.001 - (kmedoids_alternate_manhattan_random_score / 1000),
                     0.001 - (kmedoids_alternate_manhattan_heuristic_score / 1000),
                     0.001 - (kmedoids_alternate_manhattan_k_medoids_plus_score / 1000),
                     0.001 - (gmm_random_score / 1000),
                     0.001 - (gmm_kmeans_plus_score / 1000)]

#COMPUTING THE REMAINING PURITY SCORE STANDARDIZED:
purity_scores = [0.001 - (kmeans_random_lloyd_purityscore / 1000),                              #Computing the Purity Score, that will be added to the Final Comparison.
                 0.001 - (kmeans_random_purityscore / 1000), 
                 0.001 - (kmeans_random_elkan_purityscore / 1000), 
                 0.001 - (kmeans_naive_lloyd_purityscore / 1000),
                 0.001 - (kmeans_naive_elkan_purityscore / 1000),
                 0.001 - (kmeans_plus_lloyd_purityscore / 1000),
                 0.001 - (kmeans_plus_byhand_purityscore / 1000),
                 0.001 - (kmeans_plus_elkan_purityscore / 1000),
                 0.001 - (kmedians_byhand_purityscore / 1000),
                 0.001 - (kmedoids_pam_manhattan_build_purityscore / 1000),
                 0.001 - (kmedoids_alternate_manhattan_random_purityscore / 1000),
                 0.001 - (kmedoids_alternate_manhattan_heuristic_purityscore / 1000),
                 0.001 - (kmedoids_alternate_manhattan_k_medoids_plus_purityscore / 1000),
                 0.001 - (gmm_random_purityscore / 1000),
                 0.001 - (gmm_kmeans_plus_purityscore / 1000)]

#COMPUTING THE REMAINING RAND INDEX STANDARDIZED:
rand_index_score = [0.001 - (kmeans_random_lloyd_randindex / 1000),                             #Computing the Remaining Rand Index, that will be added to the Final Comparison.
                    0.001 - (kmeans_random_randindex / 1000), 
                    0.001 - (kmeans_random_elkan_randindex / 1000), 
                    0.001 - (kmeans_naive_lloyd_randindex / 1000),
                    0.001 - (kmeans_naive_elkan_randindex / 1000),
                    0.001 - (kmeans_plus_lloyd_randindex / 1000),
                    0.001 - (kmeans_plus_byhand_randindex / 1000),
                    0.001 - (kmeans_plus_elkan_randindex / 1000),
                    0.001 - (kmedians_byhand_randindex / 1000),
                    0.001 - (kmedoids_pam_manhattan_build_randindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_random_randindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_heuristic_randindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_k_medoids_plus_randindex / 1000),
                    0.001 - (gmm_random_randindex / 1000),
                    0.001 - (gmm_kmeans_plus_randindex / 1000)]

#COMPUTING THE REMAINING ADJUSTED RAND INDEX STANDARDIZED:
adj_rand_index_score = [0.001 - (kmeans_random_lloyd_adjrandindex / 1000),                      #Computing the Remaining Adjusted Rand Index, that will be added to the Final Comparison.
                    0.001 - (kmeans_random_adjrandindex / 1000), 
                    0.001 - (kmeans_random_elkan_adjrandindex / 1000), 
                    0.001 - (kmeans_naive_lloyd_adjrandindex / 1000),
                    0.001 - (kmeans_naive_elkan_adjrandindex / 1000),
                    0.001 - (kmeans_plus_lloyd_adjrandindex / 1000),
                    0.001 - (kmeans_plus_byhand_adjrandindex / 1000),
                    0.001 - (kmeans_plus_elkan_adjrandindex / 1000),
                    0.001 - (kmedians_byhand_adjrandindex / 1000),
                    0.001 - (kmedoids_pam_manhattan_build_adjrandindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_random_adjrandindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_heuristic_adjrandindex / 1000),
                    0.001 - (kmedoids_alternate_manhattan_k_medoids_plus_adjrandindex / 1000),
                    0.001 - (gmm_random_adjrandindex / 1000),
                    0.001 - (gmm_kmeans_plus_adjrandindex / 1000)]

#COMPUTING THE DAVIS-BOULDIN SCORE STANDARDIZED:
db_index_score = [kmeans_random_lloyd_dbindex / 1000,                                           #Computing the Davis-Bouldin Index, that will be added to the Final Comparison.
                  kmeans_random_dbindex / 1000, 
                  kmeans_random_elkan_dbindex / 1000,
                  kmeans_naive_lloyd_dbindex / 1000,
                  kmeans_naive_elkan_dbindex / 1000,
                  kmeans_plus_lloyd_dbindex / 1000,
                  kmeans_plus_byhand_dbindex / 1000,
                  kmeans_plus_elkan_dbindex / 1000,
                  kmedians_byhand_dbindex / 1000,
                  kmedoids_pam_manhattan_build_dbindex / 1000,
                  kmedoids_alternate_manhattan_random_dbindex / 1000,
                  kmedoids_alternate_manhattan_heuristic_dbindex / 1000,
                  kmedoids_alternate_manhattan_k_medoids_plus_dbindex / 1000,
                  gmm_random_dbindex / 1000,
                  gmm_kmeans_plus_dbindex / 1000]

#COMPUTING THE AVERAGE TIME PERFORMANCES STANDARDIZED:
avg_time = [sum(elapsed_time_kmeans['kmeans_init_lloyd']) / len(elapsed_time_kmeans['kmeans_init_lloyd']),          #Computing the average time, that will be added to the Final Comparison.
            sum(elapsed_time_kmeans['kmeans_init_byhand']) / len(elapsed_time_kmeans['kmeans_init_byhand']),
            sum(elapsed_time_kmeans['kmeans_init_elkan']) / len(elapsed_time_kmeans['kmeans_init_elkan']),
            sum(elapsed_time_kmeans['kmeans_naive_lloyd']) / len(elapsed_time_kmeans['kmeans_naive_lloyd']),
            sum(elapsed_time_kmeans['kmeans_naive_elkan']) / len(elapsed_time_kmeans['kmeans_naive_elkan']), 
            sum(elapsed_time_kmeans_plus['kmeans++_lloyd']) / len(elapsed_time_kmeans_plus['kmeans++_lloyd']),      
            sum(elapsed_time_kmeans_plus['kmeans++_byhand']) / len(elapsed_time_kmeans_plus['kmeans++_byhand']),
            sum(elapsed_time_kmeans_plus['kmeans++_elkan']) / len(elapsed_time_kmeans_plus['kmeans++_elkan']),
            sum(elapsed_time_kmedoids['kmedians']) / len(elapsed_time_kmedoids['kmedians']),        
            sum(elapsed_time_kmedoids['kmedoids_pam_build']) / len(elapsed_time_kmedoids['kmedoids_pam_build']),
            sum(elapsed_time_kmedoids['kmedoids_alternate_random']) / len(elapsed_time_kmedoids['kmedoids_alternate_random']),
            sum(elapsed_time_kmedoids['kmedoids_alternate_heuristic']) / len(elapsed_time_kmedoids['kmedoids_alternate_heuristic']),
            sum(elapsed_time_kmedoids['kmedoids_alternate_plus']) / len(elapsed_time_kmedoids['kmedoids_alternate_plus']), 
            sum(elapsed_time_gmm['mog_random']) / len(elapsed_time_gmm['mog_random']), 
            sum(elapsed_time_gmm['mog_kmeans++']) / len(elapsed_time_gmm['mog_kmeans++'])]


#DEFINITION OF THE FINAL FUNCTION WILL COMPUTE THE WHOLE COMPUTATION:
def compute_best_algorithm(avg_time, remaining_accuracies, silhouette_scores, purity_scores, rand_index_score, adj_rand_index_score, db_index_score, w_time = 0.1, w_accuracy = 0.4, w_silhouette = 0.1, w_purity = 0.1, w_randindex = 0.1, w_adj_randinxex = 0.1, w_db_index = 0.1):
    final_comparison = [round(( (avg_time[0] * w_time) + (remaining_accuracies[0] * w_accuracy) + (silhouette_scores[0] * w_silhouette) + (purity_scores[0] * w_purity) + (rand_index_score[0] * w_randindex) + (adj_rand_index_score[0] * w_adj_randinxex) + (db_index_score[0] * w_db_index)), 8),       #Add up to the average time, the remaining accuracies.
                        round(( (avg_time[1] * w_time) + (remaining_accuracies[1] * w_accuracy) + (silhouette_scores[1] * w_silhouette) + (purity_scores[1] * w_purity) + (rand_index_score[1] * w_randindex) + (adj_rand_index_score[1] * w_adj_randinxex) + (db_index_score[1] * w_db_index)), 8),
                        round(( (avg_time[2] * w_time) + (remaining_accuracies[2] * w_accuracy) + (silhouette_scores[2] * w_silhouette) + (purity_scores[2] * w_purity) + (rand_index_score[2] * w_randindex) + (adj_rand_index_score[2] * w_adj_randinxex) + (db_index_score[2] * w_db_index)), 8),
                        round(( (avg_time[3] * w_time) + (remaining_accuracies[3] * w_accuracy) + (silhouette_scores[3] * w_silhouette) + (purity_scores[3] * w_purity) + (rand_index_score[3] * w_randindex) + (adj_rand_index_score[3] * w_adj_randinxex) + (db_index_score[3] * w_db_index)), 8),
                        round(( (avg_time[4] * w_time) + (remaining_accuracies[4] * w_accuracy) + (silhouette_scores[4] * w_silhouette) + (purity_scores[4] * w_purity) + (rand_index_score[4] * w_randindex) + (adj_rand_index_score[4] * w_adj_randinxex) + (db_index_score[4] * w_db_index)), 8),       #Add up to the average time, the remaining accuracies.
                        round(( (avg_time[5] * w_time) + (remaining_accuracies[5] * w_accuracy) + (silhouette_scores[5] * w_silhouette) + (purity_scores[5] * w_purity) + (rand_index_score[5] * w_randindex) + (adj_rand_index_score[5] * w_adj_randinxex) + (db_index_score[5] * w_db_index)), 8),
                        round(( (avg_time[6] * w_time) + (remaining_accuracies[6] * w_accuracy) + (silhouette_scores[6] * w_silhouette) + (purity_scores[6] * w_purity) + (rand_index_score[6] * w_randindex) + (adj_rand_index_score[6] * w_adj_randinxex) + (db_index_score[6] * w_db_index)), 8),
                        round(( (avg_time[7] * w_time) + (remaining_accuracies[7] * w_accuracy) + (silhouette_scores[7] * w_silhouette) + (purity_scores[7] * w_purity) + (rand_index_score[7] * w_randindex) + (adj_rand_index_score[7] * w_adj_randinxex) + (db_index_score[7] * w_db_index)), 8),
                        round(( (avg_time[8] * w_time) + (remaining_accuracies[8] * w_accuracy) + (silhouette_scores[8] * w_silhouette) + (purity_scores[8] * w_purity) + (rand_index_score[8] * w_randindex) + (adj_rand_index_score[8] * w_adj_randinxex) + (db_index_score[8] * w_db_index)), 8),       #Add up to the average time, the remaining accuracies.
                        round(( (avg_time[9] * w_time) + (remaining_accuracies[9] * w_accuracy) + (silhouette_scores[9] * w_silhouette) + (purity_scores[9] * w_purity) + (rand_index_score[9] * w_randindex) + (adj_rand_index_score[9] * w_adj_randinxex) + (db_index_score[9] * w_db_index)), 8),
                        round(( (avg_time[10] * w_time) + (remaining_accuracies[10] * w_accuracy) + (silhouette_scores[10] * w_silhouette) + (purity_scores[10] * w_purity) + (rand_index_score[10] * w_randindex) + (adj_rand_index_score[10] * w_adj_randinxex) + (db_index_score[10] * w_db_index)), 8),
                        round(( (avg_time[11] * w_time) + (remaining_accuracies[11] * w_accuracy) + (silhouette_scores[11] * w_silhouette) + (purity_scores[11] * w_purity) + (rand_index_score[11] * w_randindex) + (adj_rand_index_score[11] * w_adj_randinxex) + (db_index_score[11] * w_db_index)), 8),
                        round(( (avg_time[12] * w_time) + (remaining_accuracies[12] * w_accuracy) + (silhouette_scores[12] * w_silhouette) + (purity_scores[12] * w_purity) + (rand_index_score[12] * w_randindex) + (adj_rand_index_score[12] * w_adj_randinxex) + (db_index_score[12] * w_db_index)), 8),       #Add up to the average time, the remaining accuracies.
                        round(( (avg_time[13] * w_time) + (remaining_accuracies[12] * w_accuracy) + (silhouette_scores[13] * w_silhouette) + (purity_scores[13] * w_purity) + (rand_index_score[13] * w_randindex) + (adj_rand_index_score[13] * w_adj_randinxex) + (db_index_score[13] * w_db_index)), 8),
                        round(( (avg_time[14] * w_time) + (remaining_accuracies[14] * w_accuracy) + (silhouette_scores[14] * w_silhouette) + (purity_scores[14] * w_purity) + (rand_index_score[14] * w_randindex) + (adj_rand_index_score[14] * w_adj_randinxex) + (db_index_score[14] * w_db_index)), 8)]
    return final_comparison

#PLOTTING THE RESULTS:
fig, ax = plt.subplots(figsize=(15,10))                                                                     #Creatng the figure.
width = 0.75                                                                                                #the width of the bars 
final_result = compute_best_algorithm(avg_time, remaining_accuracies, silhouette_scores, purity_scores,     #Computing the final result for all the algorithms.
                                      rand_index_score, adj_rand_index_score, db_index_score)               
result = list(zip(labels_final, final_result))                                                              #Creation of a zip object between labels and final result.
final_result = sorted(enumerate(result), key=lambda i: i[1][1], reverse = True)                             #Sorting the final list depending on the final result obtained.
values = [value[1] for i, (label, value) in enumerate(final_result)]                                        #Taking the values of the final result.
labels = [value[0] for i, (label, value) in enumerate(final_result)]                                        #Taking the labels of the final result.
ind = np.arange(len(values))                                                                                #the x locations for the groups
ax.barh(ind, values, width, color="orange")                                                                 #Plot the Horizontal Bar Representation.
ax.set_yticks(ind+width/2 - 0.35)                                                                           #Add up the y ticks.
ax.set_yticklabels(labels, minor=False)                                                                     #Set the y ticks labels.
for i, v in enumerate(values):                                                                              #For every index and value in the y list.
    ax.text(v + 0.0001, i - 0.15 , str(round(v, 6)), color='black', fontweight='bold')                      #Add up the test as labels.
plt.title('Davis Bouldin Index Score')                                                                      #Add up the Title to the Plot.
plt.xlabel('Value')                                                                                         #Add up the x-label to the Plot.
plt.ylabel('Algorithm Applied')                                                                             #Add up the y-label to the Plot.
plt.show()                                                                                                  #Show the Plot.
```
